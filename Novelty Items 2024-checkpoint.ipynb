```python
# Queries the tables from SQL
# Reads in the Weekly Sales in 2023
# Data3 Gets Guest Count
data=pd.read_sql_query('''SELECT *
  FROM [ritas_bi].[dbo].[tmp_raw_plda]
  WHERE YEAR(StartDate)=2024''',connection)
product_sales=pd.read_sql_query('''Select * from ritascoolnet_live.dbo.tblProductSalesTransactions Where Year(dteStartDate)>=2024 ''',connection)
```

```python
product_sales=product_sales.rename(columns={'dteStartDate':'StartDate'})
product_sales.intRitasStoreID=product_sales.intRitasStoreID.astype('int')
product_sales[product_sales.intRitasStoreID>10000].groupby('intRitasStoreID')['intTransactionCount'].sum()
```

```python
product_sales['Guest_Count']=product_sales.intTransactionCount-product_sales.intUECount\
-product_sales.intDDCount-product_sales.intGHCount
```

```python
test_data=data[~(data.LegacyShopType.str.contains('Mobile'))&(
    data.ProductTypeName!='Catering (Bulk Ice)')&(
    data.ProductTypeName!='Full Service Catering (Open $)')]
```

```python
gc=product_sales[product_sales.dteEndDate<=data.EndDate.max()]
gc_group=gc.groupby('intRitasStoreID')['Guest_Count'].sum().reset_index()
```

```python
test_pivot=test_data.pivot_table(index=['ShopID','ShopName'],columns='ProductTypeName',values='SalesMny',aggfunc='sum',fill_value=0,margins=True)
```

```python
novelty=test_pivot.filter(regex='Cookie|Novelty|All').iloc[0:-1,]
novelty=novelty.reset_index()
```

```python
gc_group=gc_group.rename(columns={'intRitasStoreID':'ShopID'})



combined=novelty.merge(gc_group,how='left',on='ShopID')
combined=combined[combined.All>0]
combined.info()
combined['Cookie_Sales']=combined.iloc[:,2]+combined.iloc[:,3]+combined.iloc[:,4]
combined
```

```python
combined=combined[['ShopID','ShopName','Novelty Custard (Pints)','Cookie_Sales','All','Guest_Count']]
combined
```

```python
conditions = [
    (combined["Novelty Custard (Pints)"] > 0) & (combined["Cookie_Sales"] > 0),  # Both greater than 0
    (combined["Novelty Custard (Pints)"] > 0) | (combined["Cookie_Sales"] > 0),  # One greater than 0
    (combined["Novelty Custard (Pints)"] == 0) & (combined["Cookie_Sales"] == 0),  # Both are 0
]
values = [2, 1, 0]

combined['Logic'] = np.select(conditions, values)
combined
```

```python

both=combined[combined.Logic==2]
one=combined[combined.Logic==1]
none=combined[combined.Logic==0]
```

```python
both.loc['Total']=both.iloc[:,[2,3,4,5]].sum(axis=0)
both['Check Average']=both.All/both.Guest_Count
both=both.fillna('')
both=both.drop('Logic',axis=1)
```

```python
one.loc['Total']=one.iloc[:,[2,3,4,5]].sum(axis=0)
one['Check Average']=one.All/one.Guest_Count
one=one.fillna('')
one=one.drop('Logic',axis=1)
```

```python
none.loc['Total']=none.iloc[:,[2,3,4,5]].sum(axis=0)
none['Check Average']=none.All/none.Guest_Count
none=none.fillna('')
none=none.drop('Logic',axis=1)
```

```python
with pd.ExcelWriter('Average Check Based on Novelty & Cookie Sales.xlsx',engine='xlsxwriter') as writer:
    wb=writer.book
    format_header1=wb.add_format({'bold':True,'italic':True})
    border_format=wb.add_format({'right':2,'left':2,'top':2,'bottom':2})
    whole_dollar=wb.add_format({'num_format':'$#,##0','align':'center'})
    check_format=wb.add_format({'num_format':'$#,##0.00','align':'center'})
    number_format=wb.add_format({'num_format':'#,##0','align':'center'})
    word_format=wb.add_format({'align':'center','valign':'vcenter'})
    
    text1='Ritas Franchise Company,LLC'
    text2='Check Average for Both Novelty & Cookie'
    text3='Check Average for either Novelty or Cookie'
    text4='Check Average for Neither Novelty or Cookie'
    text5='Catering and Mobile Sales Are Excluded'
    text6='Sales through : '+str(data.EndDate.max().strftime('%m-%d-%y'))
    
    sheets_data=[(both,'Novelty AND Cookies'),
                (one,'Novelty OR Cookies'),
                (none,'Neither')]
    
        # Write data to sheets and format them
    for df, sheet_name in sheets_data:
        df.to_excel(writer, startrow=5, startcol=0, sheet_name=sheet_name, index=False,header=True)
        worksheet = writer.sheets[sheet_name]
        
        # Row banding formats (for alternating colors)
        odd_row_format = wb.add_format({'bg_color': '#FFFFFF'})  # White for odd rows
        even_row_format = wb.add_format({'bg_color': '#DCE6F1'})  # Light Green for even rows

        worksheet.conditional_format(xlsxwriter.utility.xl_range(5,0,len(df)+5,len(df.columns)-1),
                                      {'type': 'no_errors', 'format': border_format})

        # Apply alternating row colors using a single conditional format
        worksheet.conditional_format(6, 0, len(df) + 5, len(df.columns) - 1, {
            'type': 'formula',
            'criteria': 'ISEVEN(ROW()-5)',
            'format': even_row_format
        })
        worksheet.conditional_format(6, 0, len(df) + 5, len(df.columns) - 1, {
            'type': 'formula',
            'criteria': 'ISODD(ROW()-5)',
            'format': odd_row_format
        })
        
        # Defines Column Header Format
        col_header=wb.add_format({'bg_color':'#8DB4E2','bold':True,'align':'center','valign':'vcenter'})
        worksheet.write_row('A6',df.columns,col_header)
        worksheet.write(0,1,text1,format_header1)
        worksheet.write(1,1,text2,format_header1)
        worksheet.write(2,1,text5,format_header1)
        worksheet.write(3,1,text6,format_header1)
        worksheet.set_column('A:A',10,word_format)
        worksheet.set_column('B:B',46, word_format)
        worksheet.set_column('C:E',22, whole_dollar)
        worksheet.set_column('F:F',12, number_format)
        worksheet.set_column('G:G',15, check_format)
        worksheet.insert_image('A1','Logo.png')
        worksheet.hide_gridlines(2)
    
```
