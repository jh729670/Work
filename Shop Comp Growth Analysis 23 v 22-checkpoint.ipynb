```python
comp_volume=total[total.StringID.isin(comp_stores)].pivot_table(index=['StringID','State','Region'],
                                                   columns='Year',
                                                   values='SalesQty',
                                                   aggfunc='sum').reset_index()
comp_volume.shape
```

```python
comp_vol=comp_volume.merge(data[['ShopID','ShopName']],left_on='StringID',right_on='ShopID').drop_duplicates(subset=[2024,2023],keep='first')
```

```python
# comp_vol=comp_vol.drop('StringID',axis=1)
comp_vol=comp_vol[['ShopID','ShopName','State','Region',2024,2023]]
comp_vol['Qty_Change %']=comp_vol[2024]/comp_vol[2023]-1
comp_vol
```

```python
comp_sales=total[total.StringID.isin(comp_stores)].pivot_table(index=['StringID','State','Region'],
                                                   columns='Year',
                                                   values='SalesMny',
                                                   aggfunc='sum').reset_index()
comp_sales.shape
```

```python
comp_s=comp_sales.merge(data[['ShopID','ShopName']],left_on='StringID',right_on='ShopID').drop_duplicates(subset=[2024,2023],keep='first')
```

```python
comp_s=comp_s.drop('StringID',axis=1)
comp_s=comp_s[['ShopID','ShopName','State','Region',2024,2023]]
comp_s['Sales_Change %']=comp_s[2024]/comp_s[2023]-1
comp_s
```

```python
comp_vol.rename(columns={2024:'Qty 24',2023:'Qty 23'},inplace=True)
comp_s.rename(columns={2024:'Sales 24',2023:'Sales 23'},inplace=True)
```

```python
comp_sv=comp_s.merge(comp_vol[['ShopID','Qty 24','Qty 23','Qty_Change %']],on='ShopID',how='left')
```

```python
comp_sv=comp_sv.drop_duplicates(subset=['ShopID'],keep='first')
comp_sv=comp_sv[['ShopID','ShopName','State','Region','Sales 24','Sales 23','Sales_Change %','Qty 24','Qty 23','Qty_Change %']]
comp_sv=comp_sv.merge(gc_piv[['ShopID',2024,2023,'GC%_Chng']],on='ShopID',how='left')
comp_sv.rename(columns={2024:'GC 24',2023:'GC 23'},inplace=True)
comp_sv
```

```python
summary=pd.DataFrame({'Statistics':['Sales 23','Sales 24','Qty 23','Qty 24','GC 23','GC 24','Sales Growth','Qty Growth','GC Growth','Weeks Open 2023','Weeks Open 2024'],
                    'Averages':[comp_sv['Sales 23'].mean(),comp_sv['Sales 24'].mean(),
                                comp_sv['Qty 23'].mean(),comp_sv['Qty 24'].mean(),comp_sv['GC 23'].mean(),comp_sv['GC 24'].mean(),comp_sv['Sales 24'].sum()/comp_sv['Sales 23'].sum()-1,
                      comp_sv['Qty 24'].sum()/comp_sv['Qty 23'].sum()-1,
                    comp_sv['GC 24'].sum()/comp_sv['GC 23'].sum()-1,
                               comp_weeks[2023].mean(),
                               comp_weeks[2024].mean()] })
```

```python
state_comps=round(comp_sv.groupby('State')[['Sales 24','Sales 23','Qty 24','Qty 23','GC 24','GC 23']].sum(),2).reset_index()
state_comps['Sales_Change %']=state_comps['Sales 24']/state_comps['Sales 23']-1
state_comps['Qty_Change %']=state_comps['Qty 24']/state_comps['Qty 23']-1
state_comps['GC_Change %']=state_comps['GC 24']/state_comps['GC 23']-1
state_comps=state_comps[['State','Sales 24','Sales 23','Sales_Change %','Qty 24','Qty 23','Qty_Change %','GC 24','GC 23','GC_Change %']]
state_comps.sort_values(by='Qty_Change %',ascending=False)
```

```python
shops_state=comp_sv.groupby('State')['ShopID'].nunique().reset_index()
shops_state.rename(columns={'ShopID':'Number_of_Shops'},inplace=True)
```

```python
state_comps=state_comps.merge(shops_state,on='State')
```

```python
writer=pd.ExcelWriter('Comp Shop State Analysis Week '+str(data.Week.max())+'.xlsx')
summary.to_excel(writer, sheet_name='Summary',index=False)
comp_sv.to_excel(writer,sheet_name='Comp Shops',index=False)
state_comps.to_excel(writer, sheet_name='State Comps',index=False)
writer.close()
```
