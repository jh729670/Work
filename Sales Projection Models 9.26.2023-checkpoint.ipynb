```python
srs_group=srs.groupby(['ShopID','intStoreID','Month','Year','SpecialVenue'])[['SalesMny','SalesQty']].sum().reset_index()
```

```python
srs_group
```

```python
gc_group=gc.groupby(['intStoreID','Month','Year'])['Guest_Count'].sum().reset_index()
```

```python
merge_df=pd.merge(srs_group,gc_group,on=['intStoreID','Month','Year'],how='left').dropna()
```

```python
merge_df['Avg_Check']=round(merge_df['SalesMny']/merge_df['Guest_Count'],2)
```

```python
le=LabelEncoder()
label=le.fit_transform(merge_df['SpecialVenue'])
label
```

```python
merge_df.drop('SpecialVenue',axis=1,inplace=True)
```

```python
merge_df['SpecialVenue']=label
merge_df.columns
```

```python
merge_df.dropna(inplace=True)
```

```python
merge_df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
train=merge_df[merge_df.Year<=2022]
test=merge_df[merge_df.Year==2023]
```

```python
x_train=train.loc[:,['Month', 'Year' ,'SalesQty',
       'Guest_Count', 'Avg_Check', 'SpecialVenue']]
y_train=train.loc[:,'SalesMny']
x_test=test.loc[:,['Month', 'Year' ,'SalesQty',
       'Guest_Count', 'Avg_Check', 'SpecialVenue']]
y_test=test.loc[:,'SalesMny']
```

```python
y_train
```

```python
#target=merge_df.loc[:,'SalesMny']
#target
```

```python
features=merge_df.loc[:,['Month', 'Year' ,'SalesQty',
       'Guest_Count', 'Avg_Check', 'SpecialVenue']]
features
```

```python
#x_train, x_test,y_train,y_test=train_test_split(features,target,test_size=.2, random_state=42)
```

```python
boosting=xgb.XGBRegressor(n_estimators=10000,early_stopping=50,learning_rate=.01)
xgb_model=boosting.fit(x_train, y_train, eval_set=[(x_test,y_test)],verbose=2000)
```

```python
#This will save our model for future use
os.chdir(r'L:\J.Harned\Models')
joblib.dump(xgb_model,'XGB Model Sales')
```

```python
fi=pd.DataFrame(data=boosting.feature_importances_,
            index=boosting.feature_names_in_,
            columns=['Importance'])
fi.sort_values('Importance').plot(kind='barh',title='Feature Importance')
```

```python
boosting.predict(x_test)
```

```python
x_test['XGB Prediction']=boosting.predict(x_test)
x_test
```

```python
test=merge_df.merge(x_test[['XGB Prediction']],how='left',left_index=True,right_index=True).dropna()
```

```python
mae_score=mean_absolute_error(test['SalesMny'],test['XGB Prediction'])
mse_score=np.sqrt(mean_squared_error(test['SalesMny'],test['XGB Prediction']))
print(f'XGB Boost Mean Absolute Error on Test Set: {mae_score:0.2f}')
print(f'XGB Boost Mean Sqaure Error on Test Set: {mse_score:0.2f}')
```
