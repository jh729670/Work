```python
import pyodbc as odbc
import pandas as pd
import os
from datetime import date,timedelta,datetime
import matplotlib.pyplot as plt
```

```python
# Names the needed driver, server name, and database
driver_name='SQL Server'
server_name='10.0.1.218'
database='ritascoolnet_live'
#Creates the connection string
connection_string=f"""
DRIVER={{{driver_name}}};
SERVER={server_name};
DATABASE={database};
Trust_Connection=yes"""
#Creates the Connection to SQL Server
connection=odbc.connect(connection_string)
print(connection)
# Queries the tables from SQL
# Reads in the Daily Sales from 2023
data1=pd.read_sql_query('''SELECT * FROM [dbo].[smartDailySales] WHERE YEAR(dteDate)>=2022''',connection)
```

```python
data1.head()
```

```python
df=data1[data1.dteDate>='2023-08-07'].pivot_table(index='vchProductSizeName',values=['intQuantity','mnyPosSales'],aggfunc='sum')
```

```python
os.chdir(r'G:\FinanceReports\2023\Wk35')
quant=df['intQuantity'].sum()
sales=df['mnyPosSales'].sum()
df['Quantity%']=df['intQuantity']/quant
df['Sales%']=df['mnyPosSales']/sales
df.to_excel('Frozen Lemonades Pmix for ' +str(data1.dteDate.max().strftime('%Y-%m-%d'))+'.xlsx')
```
