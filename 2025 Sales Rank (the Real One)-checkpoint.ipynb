```python
#Imports the necessary libraries
import pandas as pd
import os
import numpy as np
import pyodbc as odbc
from datetime import date,timedelta,datetime
import datetime as dt
import xlsxwriter
os.chdir(r'L:\J.Harned\Calendar')
calendar=pd.read_excel('2025 Sales Calendar.xlsx')
calendar['Year']=calendar.EndDate.dt.year
os.chdir(r'G:\FinanceReports\2025\Wk4')
# Names the needed driver, server name, and database
driver_name='SQL Server'
server_name='10.0.1.218'
database='ritascoolnet_live'
#Creates the connection string
connection_string=f"""
DRIVER={{{driver_name}}};
SERVER={server_name};
DATABASE={database};
Trust_Connection=yes"""
#Creates the Connection to SQL Server
connection=odbc.connect(connection_string)
print(connection)
```

```python
# Concats the two dataframes together
sales=pd.concat([data,df])

sales.loc[sales.ShopID==31360,'ShopID']=1568 #This changes the Linc's ShopID. It was changed in 2024 and at the time
                                            # of this message, 2023 sales will not line up with 2024's
sales.loc[sales.intStoreID==5444,'intStoreID']=5547
# Merges Shop Detail to the sales data frame
sales=sales.merge(data3[['intStoreID','ParentID','FC','Franchisee','SV']],on='intStoreID',how='left')
```

```python
# Creates a pivot table for Year over Year sales/guest count/ Average Check comparison
cy_pivot=cy_sales.pivot_table(index=['ParentID'],columns='Year',
                     values='SalesMny',aggfunc='sum',fill_value=0).reset_index()
gc_pivot=gc_total.pivot_table(index='ParentID',columns='Year',
                     values='Total_GC',aggfunc='sum',fill_value=0).reset_index()
gc_pivot=gc_pivot.iloc[:,[0,-1,1]]
cy_pivot['%_Chng']=cy_pivot[2025].div(cy_pivot[2024])-1
cy_pivot['%_Chng'].replace([np.inf,np.nan],0,inplace=True)
gc_pivot['%_GC']=gc_pivot[2025].div(gc_pivot[2024])-1
gc_pivot['%_GC'].replace([np.inf,np.nan],0,inplace=True)

cy_pivot=cy_pivot.merge(shop_details[['ParentID','ShopName','FC','Franchisee','State','DT','SV','StoreStatus']],on='ParentID')
cy_pivot['Rank']=cy_pivot[2025].rank(ascending=False)
cy_pivot=cy_pivot.iloc[:,[-1,0,4,7,8,9,10,5,6,2,1,3]]
cy_pivot.rename(columns={2025:'Sales_25',
                         2024:'Sales_24',
                        'DriveThru':'DT'},inplace=True)
gc_pivot.rename(columns={2025:'GC_25',
                         2024:'GC_24'},inplace=True)
cy_pivot['Franchisee'].fillna('RFC',inplace=True)
cy_pivot=cy_pivot.sort_values(by='Sales_25',ascending=False)
cy_pivot=cy_pivot.merge(gc_pivot,on='ParentID',how='left')
cy_pivot['Avg_Chk_25']=cy_pivot['Sales_25']/cy_pivot['GC_25']
cy_pivot['Avg_Chk_24']=cy_pivot['Sales_24']/cy_pivot['GC_24']
cy_pivot['Chk_%']=cy_pivot.iloc[:,-2]/cy_pivot.iloc[:,-1]-1
cy_pivot.replace([np.inf,np.nan],0,inplace=True)
cy_pivot
```

```python

# Creates the excel spreadsheet for reveiw
text1='Ritas Franchise Company,LLC'
text2='Shop Ranking Report YTD'
text3='Shop Ranking Report for Month '+str(month_name)
text4='Shop Ranking Report for Week ' + str(data.Week.max())
text5='Sales thru '+str(data.EndDate.max().strftime('%m-%d-%y'))


recap=pd.ExcelWriter('Weekly Recap for Week '+str(data.Week.max())+'v2.xlsx')
recap_workbook=recap.book

# Defines the Header Fromat
format_header2=recap_workbook.add_format()
format_header2.set_italic()
format_header2.set_bold()

#Creates the Workbook Tab
cy_pivot.to_excel(recap,
                  startrow=4,startcol=0,
                  sheet_name='YTD Sales thru '+str(data.EndDate.max().strftime('%m-%d-%y')),index=False)
#Creates teh Sheet
worksheet=recap.sheets['YTD Sales thru '+str(data.EndDate.max().strftime('%m-%d-%y'))]

#Adds the Header
worksheet.write(0,1,text1,format_header2)
worksheet.write(1,1,text2,format_header2)
worksheet.write(2,1,text5,format_header2)
num_format=recap_workbook.add_format({'num_format':'$#,##0',
                                'align':'center',
                                'valign':'center'})
chk_format=recap_workbook.add_format({'num_format':'$#,##0.00',
                                'align':'center',
                                'valign':'center'})
perc_format=recap_workbook.add_format({'num_format':'0.0%',
                                'align':'center',
                                'valign':'center'})
gc_format=recap_workbook.add_format({'num_format':'#,##0',
                                'align':'center',
                                'valign':'center'})
text_format=recap_workbook.add_format({
                                'align':'center',
                                'valign':'center'})
border_format=recap_workbook.add_format({'top':1,'bottom':1,'left':1,'right':1})
col_header=recap_workbook.add_format({'bg_color':'#548235','bold':True,'align':'center','valign':'vcenter'})

# Row banding Format(Alternate based on Odd-Even Row Number)
odd_row_format = recap_workbook.add_format({'bg_color': '#E2EFDA'})  # Light Green for odd rows
even_row_format = recap_workbook.add_format({'bg_color': '#FFFFFF'})  # White for even rows

worksheet.set_column('A:H',20,text_format)
worksheet.set_column('C:C',45,text_format)
worksheet.set_column('I:I',26,text_format)
worksheet.set_column('J:Q',13,num_format)
worksheet.set_column('M:N',13,gc_format)
worksheet.set_column('L:L',13,perc_format)
worksheet.set_column('O:O',13,perc_format)
worksheet.set_column('R:R',13,perc_format)
worksheet.set_column('P:Q',13,chk_format)
worksheet.conditional_format(xlsxwriter.utility.xl_range(4,0,len(cy_pivot)+4,len(cy_pivot.columns)-1),
                              {'type': 'no_errors', 'format': border_format})

# Apply alternating row colors using a single conditional format
worksheet.conditional_format(5, 0, len(cy_pivot) + 4, len(cy_pivot.columns) - 1, {
                'type': 'formula',
                'criteria': 'ISEVEN(ROW()-5)',
                'format': even_row_format
            })
worksheet.conditional_format(5, 0, len(cy_pivot) + 4, len(cy_pivot.columns) - 1, {
                'type': 'formula',
                'criteria': 'ISODD(ROW()-5)',
                'format': odd_row_format
            })


worksheet.insert_image('A1','Logo.png')
worksheet.hide_gridlines(2)
worksheet.autofilter(4,0,cy_pivot.shape[0],cy_pivot.shape[1]-1)
worksheet.write_row('A5',cy_pivot.columns,col_header)

cm_pivot.to_excel(recap, startrow=4, startcol=0,sheet_name='Sales thru '+str(month_name),index=False)
worksheet2=recap.sheets['Sales thru '+str(month_name)]
worksheet2.write(0,1,text1,format_header2)
worksheet2.write(1,1,text3,format_header2)
worksheet2.write(2,1,text5,format_header2)

worksheet2.set_column('A:H',20,text_format)
worksheet2.set_column('C:C',45,text_format)
worksheet2.set_column('I:I',26,text_format)
worksheet2.set_column('J:Q',13,num_format)
worksheet2.set_column('M:N',13,gc_format)
worksheet2.set_column('L:L',13,perc_format)
worksheet2.set_column('O:O',13,perc_format)
worksheet2.set_column('R:R',13,perc_format)
worksheet2.set_column('P:Q',13,chk_format)
worksheet2.conditional_format(xlsxwriter.utility.xl_range(4,0,len(cm_pivot)+4,len(cm_pivot.columns)-1),
                              {'type': 'no_errors', 'format': border_format})

worksheet2.conditional_format(5, 0, len(cm_pivot) + 5, len(cm_pivot.columns) - 1, {
                'type': 'formula',
                'criteria': 'ISEVEN(ROW()-5)',
                'format': even_row_format
            })
worksheet2.conditional_format(5, 0, len(cm_pivot) + 5, len(cm_pivot.columns) - 1, {
                'type': 'formula',
                'criteria': 'ISODD(ROW()-5)',
                'format': odd_row_format
            })

worksheet2.insert_image('A1','Logo.png')
worksheet2.hide_gridlines(2)
worksheet2.autofilter(4,0,cm_pivot.shape[0],cm_pivot.shape[1]-1)
worksheet2.write_row('A5',cm_pivot.columns,col_header)

cw_pivot.to_excel(recap,startrow=4,startcol=0, sheet_name='Sales Week '+str(data.Week.max()),index=False)
worksheet1=recap.sheets['Sales Week '+str(data.Week.max())]
worksheet1.write(0,1,text1,format_header2)
worksheet1.write(1,1,text4,format_header2)
worksheet1.write(2,1,text5,format_header2)

worksheet1.set_column('A:H',20,text_format)
worksheet1.set_column('C:C',45,text_format)
worksheet1.set_column('I:I',26,text_format)
worksheet1.set_column('J:Q',13,num_format)
worksheet1.set_column('M:N',13,gc_format)
worksheet1.set_column('L:L',13,perc_format)
worksheet1.set_column('O:O',13,perc_format)
worksheet1.set_column('R:R',13,perc_format)
worksheet1.set_column('P:Q',13,chk_format)
worksheet1.conditional_format(xlsxwriter.utility.xl_range(4,0,len(cw_pivot)+4,len(cw_pivot.columns)-1),
                              {'type': 'no_errors', 'format': border_format})

worksheet1.conditional_format(5, 0, len(cw_pivot) + 4, len(cw_pivot.columns) - 1, {
                'type': 'formula',
                'criteria': 'ISEVEN(ROW()-5)',
                'format': even_row_format
            })
worksheet1.conditional_format(5, 0, len(cw_pivot) + 4, len(cw_pivot.columns) - 1, {
                'type': 'formula',
                'criteria': 'ISODD(ROW()-5)',
                'format': odd_row_format
            })

worksheet1.insert_image('A1','Logo.png')
worksheet1.hide_gridlines(2)
worksheet1.autofilter(4,0,cw_pivot.shape[0],cw_pivot.shape[1]-1)
worksheet1.write_row('A5',cw_pivot.columns,col_header)

recap.close()
```

```python
#State Growth
core_sales=sales[(sales.Year>=2024)&(sales.State.isin(core))&(sales.Week<=data.Week.max())].pivot_table(
index='State',columns='Year',values=['SalesMny','ParentID'],aggfunc={'SalesMny':'sum','ParentID':'nunique'}).reset_index()
core_sales['Growth']=core_sales.iloc[:,-1].div(core_sales.iloc[:,-2])-1
dev_sales=sales[(sales.Year>=2024)&~(sales.State.isin(core))&(sales.Week<=data.Week.max())].pivot_table(
index='State',columns='Year',values=['SalesMny','ParentID'],aggfunc={'SalesMny':'sum','ParentID':'nunique'},fill_value=0).reset_index()
dev_sales['Growth']=dev_sales.iloc[:,-1].div(dev_sales.iloc[:,-2])-1
dev_sales.Growth.replace([np.inf,np.nan],0,inplace=True)
```

```python
#Comp Store Analysis
comp=cy_pivot[(cy_pivot.loc[:,'Sales_25']>0)&(cy_pivot.loc[:,'Sales_24']>0)]
core_comp=comp[comp.State.isin(core)]
develop_comp=comp[~comp.State.isin(core)]
core_dt_comp=core_comp[core_comp.DT=='Yes']
develop_dt_comp=develop_comp[develop_comp.DT=='Yes']
core_comp_shops=core_comp.ParentID
develop_comp_shops=develop_comp.ParentID
```

```python
#3rd Party Sales State Analysis
core_all=core_3pd.pivot_table(index='State',columns='Year',values='SalesMny',aggfunc='sum').reset_index()
core_all['Growth']=core_all[2025]/core_all[2024]-1
core_all=core_all[['State',2025,2024,'Growth']]
dev_all=dev_3pd.pivot_table(index='State',columns='Year',values='SalesMny',aggfunc='sum',fill_value=0).reset_index()
dev_all['Growth']=dev_all[2025]/dev_all[2024]-1
dev_all=dev_all[['State',2025,2024,'Growth']]
dev_all.replace([np.inf,np.nan],0,inplace=True)
#3rd Party Comp State Analysis
core_comp_3pd=core_3pd[core_3pd.ParentID.isin(core_comp_shops)].pivot_table(
    index='State',columns='Year',values='SalesMny',aggfunc='sum').reset_index()
core_comp_3pd['Growth']=core_comp_3pd[2025]/core_comp_3pd[2024]-1
core_comp_3pd=core_comp_3pd[['State',2025,2024,'Growth']]
dev_comp_3pd=dev_3pd[dev_3pd.ParentID.isin(develop_comp_shops)].pivot_table(index='State',columns='Year',
                                                                     values='SalesMny',aggfunc='sum',fill_value=0).reset_index()
dev_comp_3pd['Growth']=dev_comp_3pd[2025]/dev_comp_3pd[2024]-1
dev_comp_3pd=dev_comp_3pd[['State',2025,2024,'Growth']]
dev_comp_3pd.Growth.replace([np.inf,np.nan],0,inplace=True)
```

```python
#Comp State Growth
comp_core_sales=sales[(sales.Year>=2024)&(sales.ShopID.isin(core_comp_shops))&(sales.Week<=data.Week.max())].pivot_table(
index='State',columns='Year',values=['SalesMny','ParentID'],aggfunc={'SalesMny':'sum','ParentID':'nunique'}).reset_index()
comp_core_sales['Growth']=comp_core_sales.iloc[:,-1]/comp_core_sales.iloc[:,-2]-1
comp_core_sales
comp_dev_sales=sales[(sales.Year>=2024)&(sales.ShopID.isin(develop_comp_shops))&(sales.Week<=data.Week.max())].pivot_table(
index='State',columns='Year',values=['SalesMny','ParentID'],aggfunc={'SalesMny':'sum','ParentID':'nunique'},fill_value=0).reset_index()
comp_dev_sales['Growth']=comp_dev_sales.iloc[:,-1].div(comp_dev_sales.iloc[:,-2])-1
comp_dev_sales.Growth.replace([np.inf,np.nan],0,inplace=True)
```
