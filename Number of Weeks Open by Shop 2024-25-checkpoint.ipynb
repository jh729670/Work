```python
#Imports the necessary libraries
import pandas as pd
import os
import numpy as np
import pyodbc as odbc
from datetime import date,timedelta,datetime
import xlsxwriter
os.chdir(r'G:\FinanceReports\2025\Wk7')
```

```python
# Names the needed driver, server name, and database
driver_name='SQL Server'
server_name='10.0.1.218'
database='ritascoolnet_live'
```

```python
#Creates the connection string
connection_string=f"""
DRIVER={{{driver_name}}};
SERVER={server_name};
DATABASE={database};
Trust_Connection=yes""" 
```

```python
#Creates the Connection to SQL Server
connection=odbc.connect(connection_string)
print(connection)
```

```python
#Queries the tables from SQL
#Reads in the Daily Sales from 2023
data1=pd.read_sql_query('''SELECT * FROM [dbo].[smartDailySales] Where dteDate>= '2024-12-30' ''',connection)

data2=pd.read_sql_query('''SELECT * FROM [dbo].[smartDailySales] Where dteDate< '2024-12-30' AND dteDate>= '2024-01-01' ''',connection)

data3=pd.read_sql_query('''Select intstoreid as intStoreID,RitasID, FC,CoolNetGroup,LegacyStoreType,State,MSAName,DMA,StoreName as ShopName,CoOpName,StoreStatus from ritascoolnet_live.dbo.vShopDetail''',
                        connection)
```

```python
days=data2[data2.mnyPosSales>0].groupby(['intStoreID',pd.Grouper(key='dteDate',freq='W-SUN')])['dteDate'].nunique()\
.reset_index(name='Days_With_Sales')

weeks_open=days[days.Days_With_Sales==7].pivot_table(index=['intStoreID'],columns='Days_With_Sales',values='dteDate',aggfunc='count')


weeks_open_24=weeks_open.reset_index()
columns=['intStoreID','Number_of_Weeks24']
weeks_open_24.columns=columns
# days_open['Week'] = range(1, len(days_open) + 1)
weeks_open_24
```

```python
weeks_open_24[weeks_open_24.intStoreID==5202]
```

```python
open_shops=data3[data3.StoreStatus=='Open']
data24=data2.merge(open_shops[['intStoreID','RitasID','ShopName','FC']],on='intStoreID',how='inner')
```

```python
weeks_24=data24.groupby(pd.Grouper(key='dteDate',freq='W-SUN')).size().reset_index(name='Doesnt Matter')
weeks_24['Week']=range(1,len(weeks_24)+1)
```

```python
weeks_24
```

```python
data25=data1.merge(open_shops[['intStoreID','RitasID','ShopName','FC']],on='intStoreID',how='inner')
weeks_25=data25.groupby(pd.Grouper(key='dteDate',freq='W-SUN')).size().reset_index(name='Doesnt Matter')
weeks_25['Week']=range(1,len(weeks_25)+1)
this_week=weeks_25.Week.max()-1
this_week
```

```python
weeks_25
```

```python
all_24=data24[data24.mnyPosSales>0].groupby(['intStoreID',pd.Grouper(key='dteDate',freq='W-SUN')])['dteDate'].nunique()\
.reset_index(name='Days_With_Sales_24')
week_all_24=all_24.merge(weeks_24[['dteDate','Week']],how='left',on='dteDate')
week_all_24=week_all_24.rename(columns={'Week':'2024 YTD Weeks'})
```

```python
week_all_24
```

```python
max_week_24=week_all_24.groupby('intStoreID')['2024 YTD Weeks'].max().reset_index()
```

```python
all_25=data25[data25.mnyPosSales>0].groupby(['intStoreID',pd.Grouper(key='dteDate',freq='W-SUN')])['dteDate'].nunique()\
.reset_index(name='Days_With_Sales_25')
week_all_25=all_25.merge(weeks_25[['dteDate','Week']],how='left',on='dteDate')
week_all_25=week_all_25.rename(columns={'Week':'2025 YTD Weeks'})
```

```python
piv_24=week_all_24[(week_all_24['2024 YTD Weeks']<=this_week)&(week_all_24['Days_With_Sales_24']==7)].pivot_table(index='intStoreID',
                                                                                                          columns='Days_With_Sales_24',
                                                                                                           values='dteDate',
                                                                                                          aggfunc='nunique').reset_index()
columns=['intStoreID','2024']
piv_24.columns=columns
```

```python
piv_25=week_all_25[(week_all_25['2025 YTD Weeks']<=this_week)&(week_all_25['Days_With_Sales_25']==7)].pivot_table(index='intStoreID',
                                                                                                          columns='Days_With_Sales_25',
                                                                                                           values='dteDate',
                                                                                                          aggfunc='nunique').reset_index()
columns=['intStoreID','2025']
piv_25.columns=columns
```

```python
all_weeks=piv_25.merge(piv_24,on='intStoreID',how='outer').fillna(0)
```

```python
combined=all_weeks.merge(weeks_open_24,on='intStoreID')
combined_all=combined.merge(open_shops[['intStoreID','RitasID','ShopName','FC']],on='intStoreID',how='left')
combined_all=combined_all[['RitasID','ShopName','FC','Number_of_Weeks24','2025','2024']]
combined_all['Difference']=combined_all.iloc[:,-2]-combined_all.iloc[:,-1]
combined_all['%Difference']=combined_all.iloc[:,-1]/combined_all.iloc[:,-2]
combined_all=combined_all.replace([np.inf],0)
combined_all=combined_all.rename(columns={'Number_of_Weeks24':'2024 YE Full Weeks of Operation',
                                 '2025':'2025 YTD Full Weeks of Operation',
                                 '2024':'2024 YTD Full Weeks of Operation'})
combined_all
```

```python
combined_all[combined_all.RitasID==26]
```

```python
max_week_names_24=weeks_open_24.merge(open_shops[['intStoreID','RitasID','ShopName','FC']],on=['intStoreID'],how='left')

# max_week_names_24=max_week_names[max_week_names.RitasID.isin(combined_all.RitasID)]
max_week_names_24=max_week_names_24[['RitasID','ShopName','FC','Number_of_Weeks24']]
```

```python
max_week_names_24=max_week_names_24[(max_week_names_24.RitasID>0)&(max_week_names_24.RitasID<10000)]
# max_week_names_24=max_week_names_24.merge(combined_all[['RitasID','2024 YE Full Weeks of Operation']],on='RitasID',how='left')
# max_week_names_24=max_week_names_24[['RitasID','ShopName','FC','Number_of_Weeks24']]
max_week_names_24=max_week_names_24.rename(columns={'Number_of_Weeks24':'2024 YE Full Weeks of Operation'})
```

```python
combined_all=combined_all[combined_all.RitasID>0]
```

```python
with pd.ExcelWriter('25 v 24 Number of Shops-Full Week of Sales-WK '+str(this_week)+'.xlsx') as writer:
    combined_all.sort_values(by='RitasID').to_excel(writer,sheet_name='# of Shops',index=False,startrow=6,startcol=0)
    max_week_names_24.sort_values(by='RitasID').to_excel(writer,sheet_name='Weeks Open In 2024',startrow=6,startcol=0,index=False)
    wb=writer.book
    ws=writer.sheets['# of Shops']
    ws1=writer.sheets['Weeks Open In 2024']
    header_format=wb.add_format({'bold':True,
                                'italic':True})
    border_format=wb.add_format({'right':2,'left':2,'bottom':2,'top':2})
    perc_format=wb.add_format({'num_format':'0.0%',
                                'align':'center',
                                'valign':'vcenter'})
    wrap_text=wb.add_format({'text_wrap':True,'align':'center','valign':'vcenter'})
    # Row banding Format(Alternate based on Odd-Even Row Number)
    odd_row_format = wb.add_format({'bg_color': '#E2EFDA'})  # Light Green for odd rows
    even_row_format = wb.add_format({'bg_color': '#FFFFFF'})  # White for even rows
    
    ws.conditional_format(xlsxwriter.utility.xl_range(6,0,len(combined_all)+6,len(combined_all.columns)-1),
                              {'type': 'no_errors', 'format': border_format})
    ws.conditional_format(7, 0, len(combined_all) + 6, len(combined_all.columns) - 1, {
                'type': 'formula',
                'criteria': 'ISEVEN(ROW()-5)',
                'format': even_row_format
            })
    
    ws.conditional_format(7, 0, len(combined_all) + 6, len(combined_all.columns) - 1, {
                'type': 'formula',
                'criteria': 'ISODD(ROW()-5)',
                'format': odd_row_format
            })
    
    ws1.conditional_format(xlsxwriter.utility.xl_range(6,0,len(max_week_names_24)+6,len(max_week_names_24.columns)-1),
                              {'type': 'no_errors', 'format': border_format})
    ws1.conditional_format(7, 0, len(max_week_names_24) + 6, len(max_week_names_24.columns) - 1, {
                'type': 'formula',
                'criteria': 'ISEVEN(ROW()-5)',
                'format': even_row_format
            })
    
    ws1.conditional_format(7, 0, len(max_week_names_24) + 6, len(max_week_names_24.columns) - 1, {
                'type': 'formula',
                'criteria': 'ISODD(ROW()-5)',
                'format': odd_row_format
            })
    
    title='Ritas Franchise Company,LLC'
    text1='Full Open Shops in 2024-25 Wk '+str(this_week)
    text2='Number of Open Weeks in 2024 and 2025'
    text3='Full Open Week is 7 days of Sales Mon-Sun'
    text4='Source Code: smartDailySales'
    text5='Weeks Open in 2024'
    text6='*Special Venues Excluded'
    
    col_header=wb.add_format({'bg_color':'#548235','bold':True,'align':'center','valign':'vcenter','text_wrap':True})
    ws.write_row('A7',combined_all.columns,col_header)
    ws.write(0,1,title,header_format)
    ws.write(1,1,text1,header_format)
    ws.write(2,1,text2,header_format)
    ws.write(3,1,text3,header_format)
    ws.write(4,1,text4,header_format)
    ws.write('G6',text6,header_format)
    
    ws.set_column('A:A',12,None)
    ws.set_column('B:B',40,None)
    ws.set_column('C:C',19,None)
    ws.set_column('D:H',15,None)
    ws.set_column('H:H',15,perc_format)
    ws.insert_image('A1','Logo.png')
    ws.hide_gridlines(2)
    ws.autofilter(6,0,combined_all.shape[0],combined_all.shape[1]-1)
    
    ws1.write_row('A7',max_week_names_24.columns,col_header)
    ws1.write(0,1,title,header_format)
    ws1.write(1,1,text1,header_format)
    ws1.write(2,1,text5,header_format)
    ws1.write(3,1,text3,header_format)
    ws1.write(4,1,text4,header_format)
    ws1.write('D5',text6,header_format)
    
    ws1.set_column('A:A',12,None)
    ws1.set_column('B:B',40,None)
    ws1.set_column('C:C',19,None)
    ws1.set_column('D:H',15,None)
    ws1.insert_image('A1','Logo.png')
    ws1.hide_gridlines(2)
    ws1.autofilter(6,0,max_week_names_24.shape[0],max_week_names_24.shape[1]-1)
    
    
    
```
