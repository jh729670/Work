```python
data=pd.read_excel('2023-SRSsales-productLevel-wk20.xlsx',sheet_name='2023')
#df=pd.read_excel('2023-SRSsales-productLevel-wk19.xlsx',sheet_name='2022')
```

```python
#Creates RitasID that we can use to combine Mobile and Satellite Shops sales to their Parent Shop
data['RitasID']=data.ShopID.astype('str')
data['RitasID']=data['RitasID'].str[-4:]
data['RitasID']=data['RitasID'].astype('int')
#sorted(sales.RitasID.unique())
```

```python
# Names the needed driver, server name, and database
driver_name='SQL Server'
server_name='10.0.1.218'
database='ritascoolnet_live'
```

```python
#Creates the connection string
connection_string=f"""
DRIVER={{{driver_name}}};
SERVER={server_name};
DATABASE={database};
Trust_Connection=yes"""
```

```python
#Creates the Connection to SQL Server
connection=odbc.connect(connection_string)
print(connection)
```

```python
# Queries the tables from SQL
# Reads in the Daily Sales from 2023
data1=pd.read_sql_query('''SELECT * FROM [dbo].[smartDailySales] WHERE YEAR(dteDate)=2023''',connection)
# Reads in the Google Map to be able to join the inStoreID to the RitasID
data2=pd.read_sql_query('''Select intStoreID, intRitasStoreID, vchStoreState,vchStoreName from ritascoolnet_live.dbo.GOOGLE_MAP_store''',connection)
# Reads in the Daily Guest Count from 2023
data3=pd.read_sql_query('''SELECT * from ritascoolnet_live.dbo.smartDailyZTape Where intZTapeID=6 and Year(dteDate)=2023''',connection)
```

```python
#Filtered Guest Count on the start of the test, going 4 weeks prior to launch of Coffee Sales
filtered_gc=data3[(data3.dteDate>='2023-03-20')&(data3.dteDate<='2023-05-14')]
```

```python
filtered_gc.shape
```

```python
gc=pd.merge(filtered_gc,data2,on='intStoreID',how='left')
```

```python
gc['RitasID']=gc.intRitasStoreID.str[-4:]
gc['RitasID']=gc.RitasID.astype('int')
gc.RitasID.unique()
```

```python
gc.vchValue=gc.vchValue.astype('int')
```

```python
gc.rename(columns={'vchValue':'Guest_Count'},inplace=True)
```

```python
filtered_df=data[data.Week>=12]
```

```python
filtered_df.StartDate.min()
```

# Control Shops

```python
#Grouping control shops to get sum
control_shops=filtered_df[filtered_df.RitasID.isin(control
                                    )].groupby(['RitasID','Week'])[['SalesMny','SalesQty']].sum()
```

```python
control_pivot=control_shops.pivot_table(index='RitasID',columns='Week',values=['SalesMny','SalesQty'],aggfunc='sum',fill_value=0)
```

# Test Shops

```python
#Grouping control shops to get sum
test_shops=filtered_df[filtered_df.RitasID.isin(test
                                    )].groupby(['RitasID','Week'])[['SalesMny','SalesQty']].sum()
```

```python
test_pivot=test_shops.pivot_table(index='RitasID',columns='Week',values=['SalesMny','SalesQty'],aggfunc='sum',fill_value=0)
```

# Rogue Shops

```python
#Grouping control shops to get sum
rogue_shops=filtered_df[filtered_df.RitasID.isin(rogue
                                    )].groupby(['RitasID','Week'])[['SalesMny','SalesQty']].sum()
```

```python
rogue_pivot=rogue_shops.pivot_table(index='RitasID',columns='Week',values=['SalesMny','SalesQty'],aggfunc='sum',fill_value=0)
```

```python
rogue_pivot[['SalesMny','SalesQty']].mean().reset_index()
```

```python
os.chdir(r'L:\J.Harned\Marketing\Digital Marketing')
writer=pd.ExcelWriter('Test Data.xlsx')
control_pivot.to_excel(writer,sheet_name='Control')
test_pivot.to_excel(writer, sheet_name='Test')
rogue_pivot.to_excel(writer,sheet_name='Rogue')
writer.close()
writer.save()
```
