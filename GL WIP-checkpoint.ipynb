```python
import pandas as pd
import os
import numpy as np
import pyodbc as odbc
import datetime as dt
from datetime import date,timedelta,datetime
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# Names the needed driver, server name, and database
driver_name='SQL Server'
server_name='10.0.1.127'
database='MAS_RFC'
#Creates the connection string
connection_string=f"""
DRIVER={{{driver_name}}};
SERVER={server_name};
DATABASE={database};
Trust_Connection=yes"""
connection=odbc.connect(connection_string)
print(connection)
```

# Reads All the Necessary Tables tp Create the General Ledger

# RFC G/L

```python
#Department Names
rfc=pd.read_sql_query("""select * from mas_rfc.dbo.GL_DetailPosting where Year(PostingDate)=2024""",connection)
dept=pd.read_sql_query("""select * from mas_rfc.dbo.gl_subaccount""",connection)
# All the different accounts. Use this to determine what accounts the a JE is affecting
account=pd.read_sql_query("""select * from mas_rfc.dbo.GL_Account""",connection)
# The actual General Ledger
#Reads in the Account Group
ag=pd.read_sql_query("""select * from mas_rfc.dbo.GL_AccountGroup""",connection)
#Account Category
acc_cat=pd.read_sql_query("""select * from mas_rfc.dbo.GL_AccountCategory""",connection)
gla=rfc.merge(account[['AccountKey','AccountDesc','AccountType','AccountCategory','Account','AccountGroup']],on='AccountKey',how='left')
#Creates the Sub Account Code for the different Departments
gla['SubAccountCode']=gla.Account.str[-3:]
#Adds the Department Name
gla=gla.merge(dept[['SubAccountCode','SubAccountDesc']],on='SubAccountCode',how='left')
#Merges with the Account Group
gla_final=gla.merge(ag[['AccountType','AccountGroupDesc','AccountGroup']],on=['AccountType','AccountGroup'],how='left')
#Final Result
final_rfc=gla_final.merge(acc_cat,on='AccountCategory',how='left')
final_rfc['NetAmount']=final_rfc.DebitAmount-final_rfc.CreditAmount
```
