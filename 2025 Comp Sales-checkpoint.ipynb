```python
product_sales_24=product_sales_24.rename(columns={'dteEndDate':'EndDate'})
product_sales_25=product_sales_25.rename(columns={'dteEndDate':'EndDate'})
product_sales=pd.concat([product_sales_24,product_sales_25])
weeks=all_data[['EndDate','Week']].drop_duplicates(keep='first')
ps=product_sales.merge(weeks[['EndDate','Week']],how='left',on='EndDate')
ps['Retail_GC']=ps.intTransactionCount-ps.intUECount\
-ps.intDDCount-ps.intGHCount
threepd=ps.rename(columns={'intRitasStoreID':'ShopID'})
threepd.ShopID=threepd.ShopID.astype('int')
threepd=threepd.rename(columns={'intTransactionCount':'Total_GC'})
gc=threepd.groupby(['ShopID','Year','Week','EndDate'])['Total_GC'].sum().reset_index()
gc.Week=gc.Week.astype('int')
gc.loc[gc.ShopID==31360,'ShopID']=1568
gc['ParentID']=gc.ShopID.astype('int')
gc['ParentID']=gc.ShopID.astype('str')
gc['ParentID']=gc.ParentID.str[-4:]
gc['ParentID']=gc.ParentID.astype('int')
gc_total=gc
gc_total=gc_total[gc_total.ParentID!=491]
```

```python
lastrow=monthly_comps.shape[0]
with pd.ExcelWriter('Comp Sales and Guest Count for Wk'+str(this_week)+' v2.xlsx',engine='xlsxwriter') as writer5:
    comp_workbook=writer5.book
    format_header2=comp_workbook.add_format()
    format_header2.set_italic()
    format_header2.set_bold()
    format_num=comp_workbook.add_format({'num_format':'$#,##0',
                                        'align':'center',
                                        'valign':'vcenter'})
    format_qty=comp_workbook.add_format({'num_format':'#,##0',
                                        'align':'center',
                                        'valign':'vcenter'})
    perc_format=comp_workbook.add_format({'num_format':'#0.0%',
                                         'align':'center',
                                         'valign':'vcenter'})
    total_dollar_format=comp_workbook.add_format({'bold':True,
                                          'bg_color':'#d5d8dc',
                                          'num_format':'$#,##0',
                                        'align':'center',
                                        'valign':'vcenter'})
    
    total_gc_format=comp_workbook.add_format({'bold':True,
                                          'bg_color':'#d5d8dc',
                                          'num_format':'#,##0',
                                        'align':'center',
                                        'valign':'vcenter'})
    
    total_perc_format=comp_workbook.add_format({'bold':True,
                                          'bg_color':'#d5d8dc',
                                          'num_format':'#0.0%',
                                        'align':'center',
                                        'valign':'vcenter'})
    a1=comp_workbook.add_format({'align':'center','valign':'vcenter','bold':True})
    border_format=comp_workbook.add_format({'top':2,'bottom':2,'left':2,'right':2})
    
    disclaimer_format=comp_workbook.add_format({'bold':True,'italic':True,'text_wrap':True,'font_size':10,'valign':'vcenter'})
    
    title='Ritas Franchise Company,LLC'
    comp_text='Weekly Same Store Sales and Guest Count'
    comp_text0='Monthly Same Store Sales and Guest Count'
    comp_text1='Same Store Sales Include: Retail/Mobile/Satellite/3rd Party/Special Venue'
    comp_text2='2025 v 2024 Sales and Guest Counts thru Week '+str(this_week)
    comp_text3='As of '+str(data.EndDate.max().strftime('%m-%d-%y'))
    disclaimer_1='* For a Shop to be eligible on the Monthly Same Store Sales and Guest Count report, a shop must have had sales in the Current Year AND in the Prior Year, regardless of when the sales occurred. Note that these totals WILL NOT match the totals in the Wkly Stats tab'
    disclaimer_2='* For a Shop to be eligible on the Weekly Same Store Sales and Guest Count report, a shop must have had sales in the Current Year AND Prior Year in the Same Week. These numbers WILL BE different than the Monthly Comps, because shops are not always open during the same week Year over Year'
    monthly_comps.to_excel(writer5,startrow=6,startcol=0, sheet_name='Comps')
    true_comps.T.reset_index().to_excel(writer5,startrow=6,startcol=0, sheet_name='''Wkly Stats''',index=False)
    
    ws=writer5.sheets['Comps']                     
    ws.conditional_format(xlsxwriter.utility.xl_range(6,0,len(monthly_comps)+6,len(monthly_comps.columns)),
                                  {'type': 'no_errors', 'format': border_format})
    for col in range(0,2):
        ws.write(monthly_comps.shape[0]+6,col+1,monthly_comps.iloc[-1,col],total_dollar_format)
  
    for col in range(3,5):
        ws.write(monthly_comps.shape[0]+6,col+1,monthly_comps.iloc[-1,col],total_gc_format)
    for col in range(2,3):
        ws.write(monthly_comps.shape[0]+6,col+1,monthly_comps.iloc[-1,col],total_perc_format)
    for col in range(5,6):
        ws.write(monthly_comps.shape[0]+6,col+1,monthly_comps.iloc[-1,col],total_perc_format)
    ws.merge_range('A25:G25',disclaimer_1,disclaimer_format)
    ws.set_row(24,50)
    col_header=comp_workbook.add_format({'bg_color':'#00B0F0','bold':True,'align':'center','valign':'vcenter','text_wrap':True})
    ws.write_row('A7',monthly_comps.reset_index().columns,col_header) 
    ws.write(lastrow+6,0,'YTD Totals',total_dollar_format)    
    ws.write(0,1,title,format_header2)
    ws.write(1,1,comp_text0,format_header2)
    ws.write(2,1,comp_text1,format_header2)
    ws.write(3,1,comp_text2,format_header2)
    ws.write(4,1,comp_text3,format_header2)
    ws.set_column('A:A',13)
    ws.set_column('B:C',22,format_num)
    ws.set_column('D:D',22,perc_format)
    ws.set_column('E:F',15,format_qty)
    ws.set_column('G:G',15,perc_format)
    ws.insert_image('A1','Logo.png')
    ws.hide_gridlines(2)
   
    ws1=writer5.sheets['''Wkly Stats''']
    ws1.conditional_format(xlsxwriter.utility.xl_range(6,0,len(true_comps.T)+6,len(true_comps.T.columns)),
                                  {'type': 'no_errors', 'format': border_format})
    ws1.merge_range('A16:G16',disclaimer_2,disclaimer_format)
    ws1.set_row(15,50)
    ws1.write(0,1,title,format_header2)
    ws1.write(1,1,comp_text,format_header2)
    ws1.write(2,1,comp_text1,format_header2)
    ws1.write(3,1,comp_text2,format_header2)
    ws1.write(4,1,comp_text3,format_header2)
    ws1.set_column('A:ZZ',13)
    ws1.write('A7','Week',a1)
    ws1.set_column('A:A',23)
    ws1.set_column('B:AZ',14)
    ws1.set_row(7,None,format_num)
    ws1.set_row(8,None,format_num)
    ws1.set_row(9,None,perc_format)
    ws1.set_row(10,None,format_qty)
    ws1.set_row(11,None,format_qty)
    ws1.set_row(12,None,perc_format)
    ws1.insert_image('A1','Logo.png')
    ws1.hide_gridlines(2)
```
