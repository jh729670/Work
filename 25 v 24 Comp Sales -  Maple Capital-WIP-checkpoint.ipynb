```python
def get_initials(name):
    if not isinstance(name, str) or name.strip() == "":  # Handle None or empty strings
        return name  # Keep None or empty as is
    parts = name.split()
    if len(parts) == 1:  # If only one name, return it as is
        return name
    else:
        return ''.join([part[0].upper() for part in parts])  # Extract initials

data3['FC Initials'] = data3['FC'].apply(get_initials)
data3.loc[data3.FC=='Joe Toocheck','FC Initials']='JToo'
data3.loc[data3.FC=='John Martin (FC)','FC Initials']='JM'
```
