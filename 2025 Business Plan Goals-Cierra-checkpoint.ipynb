```python
total=df.pivot_table(index=['ParentID'],
                                                                                  columns='Year',
                                                                                  values='SalesMny',
                                                                                  aggfunc='sum',
                                                                                  fill_value=0).reset_index()
total['% Chg']=total.iloc[:,-1]/total.iloc[:,-2]-1
total['% Chg']=total['% Chg'].fillna(0)
total['% Chg']=total['% Chg'].replace([np.inf,np.nan],0)
total.head()
```

```python
alist=['3rd Party Delivery','Full Service Catering (Open $)','Online Ordering','Catering (Bulk Ice)']
retail=df[~(df.LegacyShopType.str.contains('Mobile'))&~(
    df.ProductTypeName.isin(alist))&(df.LegacyShopStatus!='Satellite Store')].pivot_table(index=['ParentID'],
                                                                                  columns='Year',
                                                                                  values='SalesMny',
                                                                                  aggfunc='sum',
                                                                                  fill_value=0).reset_index()
retail['% Chg']=retail.iloc[:,-1]/retail.iloc[:,-2]-1
retail['% Chg']=retail['% Chg'].fillna(0)
retail['% Chg']=retail['% Chg'].replace([np.inf,np.nan],0)
```

```python
digital_list=['3rd Party Delivery','Online Ordering']
digital=df[(df.ProductTypeName.isin(digital_list))&(df.SpecialVenue=='No')].pivot_table(index='ParentID',
                                                                                                          columns='Year',
                                                                                                          values='SalesMny',
                                                                                                          aggfunc='sum',
                                                                                                          fill_value=0)\
.reset_index()
digital['% Chg']=digital.iloc[:,-1]/digital.iloc[:,-2]-1
digital['% Chg']=digital['% Chg'].fillna(0)
digital['% Chg']=digital['% Chg'].replace([np.inf,np.nan],0)
digital.head(10)
```

```python
catering=df[(df.SpecialVenue=='No')&(
    df.ProductTypeName=='Full Service Catering (Open $)')|(df.ProductTypeName=='Catering (Bulk Ice)')].pivot_table(index='ParentID',
                                                                                                        columns='Year',
                                                                                                        values='SalesMny',
                                                                                                        aggfunc='sum',
                                                                                                        fill_value=0).reset_index()
catering['% Chg']=catering.iloc[:,-1]/catering.iloc[:,-2]-1
catering['% Chg']=catering['% Chg'].fillna(0)
catering['% Chg']=catering['% Chg'].replace([np.inf,np.nan],0)
catering.head(10)
```

```python
mobile=df[(df.LegacyShopType.str.contains('Mobile'))&(df.SpecialVenue=='No')&(
    df.ProductTypeName!='Full Service Catering (Open $)')].pivot_table(index='ParentID',
                                                                                                        columns='Year',
                                                                                                        values='SalesMny',
                                                                                                        aggfunc='sum',
                                                                                                        fill_value=0).reset_index()
mobile['% Chg']=mobile.iloc[:,-1]/mobile.iloc[:,-2]-1
mobile['% Chg']=mobile['% Chg'].fillna(0)
mobile['% Chg']=mobile['% Chg'].replace([np.inf,np.nan],0)
mobile.head()
```

```python
satellite=df[(df.LegacyShopStatus=='Satellite Store')&~(df.LegacyShopType.str.contains('Mobile'))].pivot_table(index=['ParentID'],
                                                                                  columns='Year',
                                                                                  values='SalesMny',
                                                                                  aggfunc='sum',
                                                                                  fill_value=0).reset_index()
satellite['% Chg']=satellite.iloc[:,-1]/satellite.iloc[:,-2]-1
satellite['% Chg']=satellite['% Chg'].fillna(0)
satellite['% Chg']=satellite['% Chg'].replace([np.inf,np.nan],0)
```

```python
ps=ps.rename(columns={'intRitasStoreID':'ShopID'})
gc=ps.groupby(['ShopID','Year'])['Retail_GC'].sum().reset_index()
gc.loc[gc.ShopID==31360,'ShopID']=1568
gc['ParentID']=gc.ShopID.astype('int')
gc['ParentID']=gc.ShopID.astype('str')
gc['ParentID']=gc.ParentID.str[-4:]
gc['ParentID']=gc.ParentID.astype('int')
gc_total=gc
gc_total=gc_total[gc_total.ParentID!=491]
gc=gc_total.pivot_table(index='ParentID',values='Retail_GC',columns='Year',aggfunc='sum',fill_value=0)
gc['% Chg']=gc.iloc[:,-1]/gc.iloc[:,-2]-1
gc['% Chg']=gc['% Chg'].fillna(0)
gc['% Chg']=gc['% Chg'].replace([np.inf,np.nan],0)
gc=gc.reset_index()
```

```python
sales_gc=retail[['ParentID',2024,2023]].merge(gc[['ParentID',2023,2024]],on='ParentID',how='left',suffixes=['_Sales','_GC'])
sales_gc['Avg_Check_24']=sales_gc['2024_Sales']/sales_gc['2024_GC']
sales_gc['Avg_Check_23']=sales_gc['2023_Sales']/sales_gc['2023_GC']
sales_gc['Check_%']=(sales_gc['Avg_Check_24']/sales_gc['Avg_Check_23']-1).fillna(0)
sales_gc['Check_%']
```
