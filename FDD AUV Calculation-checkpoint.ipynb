```python
# Reads in the Google Map to be able to join the inStoreID to the RitasID
data2=pd.read_sql_query('''Select intStoreID, intRitasStoreID as StringID, vchStoreState as State,vchStoreName as ShopName,dteOpenDate,dteTransferDate,dteSeasonalOpenDate,dteSeasonalCloseDate,bitDriveThru as DriveThru from ritascoolnet_live.dbo.GOOGLE_MAP_store''',connection)
# Pulls Marketing info needed like Coolnet Group, MSA, and DMA
data3=pd.read_sql_query('''Select intstoreid as intStoreID,RitasID as ShopID, FC,CoolNetGroup,LegacyStoreType,State,MSAName,DMA,StoreName as ShopName,CoOpName,Seasonality,StoreStatus,OpenDate from ritascoolnet_live.dbo.vShopDetail''',connection)
```

```python
# Removes shops that never opened or have closed
data2.dropna(subset='dteOpenDate',inplace=True)
data2.dropna(subset='dteSeasonalOpenDate',inplace=True)
```

```python
shop_detail=data3[data3.ShopID>0].sort_values(by=['ShopID','StoreStatus']).drop_duplicates(subset=['ShopID'],keep='first')
```

```python
#Removes duplicate shop values
shops=data2.sort_values('intStoreID').drop_duplicates(
    subset='StringID',keep='last')
```

```python
data3[data3.ShopID==1360]
```

```python
# #Converts the ShopID to numeric
shops['StringID']=pd.to_numeric(shops['StringID'],errors='coerce')
```

```python
#Aggregates the Sales Data to get sum of sales and number of unique weeks opened
df_group=data.groupby(['StringID','Region'])[['SalesMny','Week']].agg(
    {'SalesMny':'sum','Week':'nunique'}).reset_index()
```

```python
dates_group=df_group.merge(shop_detail[['intStoreID','OpenDate','ShopName','ShopID','LegacyStoreType','Seasonality',
                                       'StoreStatus']],left_on='StringID',right_on='ShopID')
```

```python
season_weeks=data[(data.EndDate>='2024-03-01')&(data.StartDate<='2024-09-30')].groupby('StringID')['Week'].nunique().reset_index()
season_weeks.rename(columns={'Week':'In_Season_Weeks'},inplace=True)
```

```python
sorted(data[(data.EndDate>='2024-03-01')&(data.StartDate<='2024-09-30')].EndDate.unique())
```

```python
totals=grouped.merge(season_weeks, on='StringID',how='left')
totals['In_Season_Weeks']=totals.In_Season_Weeks.fillna(0)
```

```python
totals.info()
```

```python
totals.SalesMny.sum()
```

```python
totals=totals.iloc[:,[7,6,1,10,8,11,2,3,13,5,12,9]]
```

```python
totals.sort_values('ShopID')
```

```python
totals.rename(columns={'Week':'Weeks_Open'},inplace=True)
```

```python
removed=totals[(totals.StoreStatus!='Open')|(totals.In_Season_Weeks<26)]
totals[~totals.ShopID.isin(removed.ShopID)]
```

```python
totals['Rank']=totals.SalesMny.rank(ascending=False)
totals=totals.iloc[:,np.r_[-1,0:len(totals.columns)-1]]
totals=totals.sort_values(by='Rank')
```

```python
final=totals[~totals.ShopID.isin(removed.ShopID)]
final=final.drop('Rank',axis=1)
final['Rank']=final.SalesMny.rank(ascending=False)
final=final.iloc[:,np.r_[-1,0:len(final.columns)-1]]
final
```
