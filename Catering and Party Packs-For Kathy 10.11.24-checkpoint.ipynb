```python
import pandas as pd
import os
import numpy as np
import pyodbc as odbc
import datetime as dt
from datetime import date,timedelta,datetime
import matplotlib.pyplot as plt
import seaborn as sns
import xlsxwriter
os.chdir(r'G:\FinanceReports\2024\Another Kathy Request')
```

```python
#Names the needed driver, server name, and database
driver_name='SQL Server'
server_name='10.0.1.218'
database='ritascoolnet_live'
#Creates the connection string
connection_string=f"""
DRIVER={{{driver_name}}};
SERVER={server_name};
DATABASE={database};
Trust_Connection=yes"""
#Creates the Connection to SQL Server
connection=odbc.connect(connection_string)
print(connection)
# Queries the tables from SQL
# Reads in the Weekly Sales in 2023
data=pd.read_sql_query('''SELECT *
  FROM [ritas_bi].[dbo].[tmp_raw_plda]
  WHERE YEAR(StartDate)>=2022
  ORDER BY Week DESC''',connection)
data2=pd.read_sql_query('''SELECT *
  FROM [ritas_bi].[dbo].[tmp_raw_pldb]
  WHERE YEAR(StartDate)=2021''',connection)
data3=pd.read_sql_query('''Select intstoreid as intStoreID,RitasID as ShopID, FC,CoolNetGroup,LegacyStoreType,State,MSAName,DMA,StoreName as ShopName,StoreStatus,[Special Venue] as SV,[Parent Shop #] as CustomerNo from ritascoolnet_live.dbo.vShopDetail''',connection)
```

```python
os.chdir(r'G:\FinanceReports\YTD Files')
twenty=pd.read_excel('2020-SRSsales-productLevel-wk1-53.xlsx',sheet_name='2020')
nineteen=pd.read_excel('2019-SRSsales-productLevel-2.xlsx',sheet_name='2019')
```

```python
data['Year']=data.StartDate.dt.year
data2['Year']=data2.StartDate.dt.year
twenty['Year']=2020
nineteen['Year']=2019
nineteen.drop(['OpenDate','FiscalMonth'],axis=1,inplace=True)
```

```python
all_sales=pd.concat([data,data2,twenty,nineteen])
all_sales.loc[all_sales.ShopID==31360,'ShopID']=1568
all_sales.loc[all_sales.ShopID==1568,'SpecialVenue']='Yes'
```

```python
all_sales_group=all_sales.groupby('Year')['SalesMny'].sum().reset_index()
all_sales_group
```

```python
open_shops=data[data.Year==2024].ShopID.unique()
open_sales=all_sales[all_sales.ShopID.isin(open_shops)]
open_sales.groupby('Year')['SalesMny'].sum().reset_index()
```

```python
open_sales.columns
```

```python
open_sales[['ProductCode','ProductSizeName']].drop_duplicates(keep='first',subset=['ProductCode','ProductSizeName']).to_excel('ProductSizeName.xlsx',index=False)
```

```python
cater=['Full Service Catering (Open $)','Catering (Bulk Ice)']
catering=open_sales[(open_sales.SpecialVenue=='No')&(
    open_sales.ProductTypeName.isin(cater))&(open_sales.ShopID.isin(open_shops))].pivot_table(index='ShopID',
                                                                                                        columns=['Year','ProductTypeName'],
                                                                                                        values='SalesMny',
                                                                                                        aggfunc='sum',
                                                                                                        fill_value=0).reset_index()
catering.sum()
```

```python
catering.columns=catering.columns.map(lambda x: '_'.join([str(i) for i in x]))
catering=catering.rename(columns={'ShopID_':'ShopID'})
catering
```

```python
shop_names=data[data.Year==2024][['ShopID','ShopName']].drop_duplicates(subset='ShopID',keep='last')
catering_shops=shop_names.merge(catering,on='ShopID',how='inner')
with pd.ExcelWriter('Catering Sales by Shop 2019-2024.xlsx') as writer1:
    catering_shops.sort_values(by='ShopID').to_excel(writer1,sheet_name='Catering Sales',index=False,startrow=5,startcol=0)
    
    wb=writer1.book
    dollar_format=wb.add_format({'num_format':'$#,##0',
                                'align':'center',
                                'valign':'vcenter'})
    header_format=wb.add_format({'bold':True,
                                'italic':True})
    border_format=wb.add_format({'right':2,'left':2,'bottom':2,'top':2})
    word_format=wb.add_format({'align':'center','valign':'vcenter'})
    
    ws=writer1.sheets['Catering Sales']
    ws.conditional_format(xlsxwriter.utility.xl_range(5,0,len(catering_shops)+5,len(catering_shops.columns)-1),
                              {'type': 'no_errors', 'format': border_format})
    
    title='Ritas Franchise Company,LLC'
    text1='Catering Sales by Product Size Name'
    text2='2019 thru '+str(data.EndDate.max().strftime('%m-%d-%y'))
    text3='Source Code: Product Level Detail'
    
    ws.write(0,1,title,header_format)
    ws.write(1,1,text1,header_format)
    ws.write(2,1,text2,header_format)
    ws.write(3,1,text3,header_format)
    ws.set_column('C:BB',12, dollar_format)
    ws.set_column('A:A',12,word_format)
    ws.set_column('B:B',40,word_format)
    ws.hide_gridlines(2)
    ws.insert_image('A1','Logo.png')
```

```python
items=[1198,1200,1201]
these_guys=all_sales[(all_sales.Year>=2022)&(all_sales.ProdSizeCode.isin(items))].pivot_table(
    index='ShopID',columns=['Year','ProductSizeName'],values=['SalesMny','SalesQty'],aggfunc='sum',fill_value=0).reset_index()
these_guys.columns=these_guys.columns.map(lambda x: ' '.join([str(i) for i in x]))
these_guys=these_guys.rename(columns={'ShopID  ':'ShopID'})
```

```python
these_guys.columns
```

```python
gallon=shop_names.merge(these_guys, on='ShopID',how='inner')
with pd.ExcelWriter('Party Packs Sales and Quantity by Shop 2022-2024.xlsx') as writer1:
    gallon.sort_values(by='ShopID').to_excel(writer1,sheet_name='Party Packs',index=False,startrow=5,startcol=0)
    
    wb=writer1.book
    dollar_format=wb.add_format({'num_format':'$#,##0',
                                'align':'center',
                                'valign':'vcenter'})
    qty_format=wb.add_format({'num_format':'#,##0','align':'center','valign':'vcenter'})
    header_format=wb.add_format({'bold':True,
                                'italic':True})
    border_format=wb.add_format({'right':2,'left':2,'bottom':2,'top':2})
    word_format=wb.add_format({'align':'center','valign':'vcenter'})
    
    ws=writer1.sheets['Party Packs']
    ws.conditional_format(xlsxwriter.utility.xl_range(5,0,len(gallon)+5,len(gallon.columns)-1),
                              {'type': 'no_errors', 'format': border_format})
    
    title='Ritas Franchise Company,LLC'
    text1='Party Sales and Quantity'
    text2='2022 thru '+str(data.EndDate.max().strftime('%m-%d-%y'))
    text3='Source Code: Product Level Detail'
    
    ws.write(0,1,title,header_format)
    ws.write(1,1,text1,header_format)
    ws.write(2,1,text2,header_format)
    ws.write(3,1,text3,header_format)
    ws.set_column('C:J',12, dollar_format)
    ws.set_column('K:Z',12, qty_format)
    ws.set_column('A:A',12,word_format)
    ws.set_column('B:B',40,word_format)
    ws.hide_gridlines(2)
    ws.insert_image('A1','Logo.png')
```
