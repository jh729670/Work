```python
import pandas as pd
import os
import numpy as np
import pyodbc as odbc
import datetime as dt
from datetime import date,timedelta,datetime
import matplotlib.pyplot as plt
import seaborn as sns
import xlsxwriter
os.chdir(r'G:\FinanceReports\2024\Another Kathy Request')
# plt.style.use('ggplot')
```

```python
#Names the needed driver, server name, and database
driver_name='SQL Server'
server_name='10.0.1.218'
database='ritascoolnet_live'
#Creates the connection string
connection_string=f"""
DRIVER={{{driver_name}}};
SERVER={server_name};
DATABASE={database};
Trust_Connection=yes"""
#Creates the Connection to SQL Server
connection=odbc.connect(connection_string)
print(connection)
# Queries the tables from SQL
# Reads in the Weekly Sales in 2023
data=pd.read_sql_query('''SELECT *
  FROM [ritas_bi].[dbo].[tmp_raw_plda]
  WHERE YEAR(StartDate)>=2022
  ORDER BY Week DESC''',connection)
data2=pd.read_sql_query('''SELECT *
  FROM [ritas_bi].[dbo].[tmp_raw_pldb]
  WHERE YEAR(StartDate)=2021''',connection)
data3=pd.read_sql_query('''Select intstoreid as intStoreID,RitasID as ShopID, FC,CoolNetGroup,LegacyStoreType,State,MSAName,DMA,StoreName as ShopName,StoreStatus,[Special Venue] as SV,[Parent Shop #] as CustomerNo from ritascoolnet_live.dbo.vShopDetail''',connection)
```

```python
os.chdir(r'L:\07 Sales Analysis\Reported Sales\Final SRS by Product by Year')
seventeen=pd.read_excel('2017 SRS product Level.xlsx',sheet_name='2017')
eighteen=pd.read_excel('2018 SRS product Level.xlsx',sheet_name='2018')
os.chdir(r'G:\FinanceReports\YTD Files')
twenty=pd.read_excel('2020-SRSsales-productLevel-wk1-53.xlsx',sheet_name='2020')
nineteen=pd.read_excel('2019-SRSsales-productLevel-2.xlsx',sheet_name='2019')
```

```python
data['Year']=data.StartDate.dt.year
data2['Year']=data2.StartDate.dt.year
twenty['Year']=2020
nineteen['Year']=2019
eighteen['Year']=2018
seventeen['Year']=2017
nineteen.drop(['OpenDate','FiscalMonth'],axis=1,inplace=True)
```

```python
all_sales=pd.concat([data,data2,twenty,nineteen,eighteen,seventeen])
all_sales.loc[all_sales.ShopID==31360,'ShopID']=1568
all_sales.loc[all_sales.ShopID==1568,'SpecialVenue']='Yes'
```

```python
open_shops=data[data.Year==2024].ShopID.unique()
```

```python
product_sales=pd.read_sql_query('''Select * from ritascoolnet_live.dbo.tblProductSalesTransactions Where Year(dteStartDate)>=2017 ''',connection)
```

```python
ps=product_sales.rename(columns={'dteStartDate':'StartDate'})
ps['Retail_GC']=ps.intTransactionCount-ps.intUECount\
-ps.intDDCount-ps.intGHCount
ps['Year']=ps.StartDate.dt.year
threepd=ps.rename(columns={'intRitasStoreID':'ShopID'})
threepd.ShopID=threepd.ShopID.astype('int')
gc=threepd.groupby(['ShopID','Year','StartDate'])['Retail_GC'].sum().reset_index()
gc.loc[gc.ShopID==31360,'ShopID']=1568
gc['ParentID']=gc.ShopID.astype('int')
gc['ParentID']=gc.ShopID.astype('str')
gc['ParentID']=gc.ParentID.str[-4:]
gc['ParentID']=gc.ParentID.astype('int')
gc_total=gc
gc_total=gc_total[gc_total.ParentID!=491]
gc_total
```

```python
gc=gc_total[gc_total.ShopID.isin(open_shops)].pivot_table(index='ShopID',values='Retail_GC',columns='Year',aggfunc='sum',fill_value=0)
```

```python
gc
```

```python
ps3=product_sales.rename(columns={'dteStartDate':'StartDate'})
ps3['3rd Party']=ps.intUECount+ps.intDDCount+ps.intGHCount
ps3['Year']=ps3.StartDate.dt.year
threepd=ps3.rename(columns={'intRitasStoreID':'ShopID'})
threepd.ShopID=threepd.ShopID.astype('int')
gc3=threepd.groupby(['ShopID','Year','StartDate'])['3rd Party'].sum().reset_index()
gc3.loc[gc3.ShopID==31360,'ShopID']=1568
gc3['ParentID']=gc3.ShopID.astype('int')
gc3['ParentID']=gc3.ShopID.astype('str')
gc3['ParentID']=gc3.ParentID.str[-4:]
gc3['ParentID']=gc3.ParentID.astype('int')
gc_total3=gc3
gc_total3=gc_total3[gc_total3.ParentID!=491]
gc_total3
```

```python
gc3=gc_total3[gc_total3.ShopID.isin(open_shops)].pivot_table(index='ShopID',values='3rd Party',columns='Year',aggfunc='sum',fill_value=0)
gc3
```
