```python
df=pd.read_sql_query('''SELECT *
  FROM [ritas_bi].[dbo].[tmp_raw_plda]
  WHERE YEAR(StartDate)>=2023
  ORDER BY Week DESC''',connection)
```

```python
df=df.merge(data3[['ShopID','DMA','MSAName','SeasonClose','SV']],how='left',on='ShopID').drop_duplicates(keep='first')
```

```python
os.chdir(r'M:\Natalie Florida 2023 PMIX\2024 PMIX')
fl=df[df.State=='FL'].ShopID.unique()
df['Year']=df.StartDate.dt.year
week_max=df[df.Year==df.Year.max()].Week.max()
for i in list(enumerate(fl)):
    fl_qty=df[(df.ShopID==i[1])&(df.Week.between(9,week_max))].SalesQty.sum()
    qty_cat=df[(df.ShopID==i[1])&(df.Week.between(9,week_max))].pivot_table(index='ProductTypeName',columns=['Week','Year'],
                                               values='SalesQty',aggfunc='sum',fill_value=0).reset_index()
    qty_item=df[(df.ShopID==i[1])&(df.Week.between(9,week_max))].pivot_table(index='ProductSizeName',columns=['Week','Year'],
                                               values='SalesQty',aggfunc='sum',fill_value=0).reset_index()
    pqty_cat=df[(df.ShopID==i[1])&(df.Week.between(9,week_max))].pivot_table(index='ProductTypeName',columns=['Week','Year'],
                                               values='SalesQty',aggfunc='sum',fill_value=0).apply(lambda x:x/x.sum()).reset_index()
    pqty_item=df[(df.ShopID==i[1])&(df.Week.between(9,week_max))].pivot_table(index='ProductSizeName',columns=['Week','Year'],
                                               values='SalesQty',aggfunc='sum',fill_value=0).apply(lambda x:x/x.sum()).reset_index()
    writer=pd.ExcelWriter('FL Shop '+str(i[1])+' PMIX 24 v 23.xlsx',engine='xlsxwriter')
    qty_cat.sort_values(by=qty_cat.columns[-1],ascending=False).to_excel(writer, sheet_name='Qty Sold Category Shop '+ str(i[1]))
    pqty_cat.sort_values(by=pqty_cat.columns[-1],ascending=False).to_excel(writer, sheet_name='Qty% Sold Category Shop '+ str(i[1]))
    qty_item.sort_values(by=qty_item.columns[-1],ascending=False).to_excel(writer, sheet_name='Qty Sold Item Shop '+ str(i[1]))
    pqty_item.sort_values(by=pqty_item.columns[-1],ascending=False).to_excel(writer, sheet_name='Qty% Sold Item Shop '+ str(i[1]))
    workbook=writer.book
    worksheet=writer.sheets['Qty Sold Category Shop '+str(i[1])]
    worksheet_1=writer.sheets['Qty% Sold Category Shop '+str(i[1])]
    worksheet_2=writer.sheets['Qty Sold Item Shop '+str(i[1])]
    worksheet_3=writer.sheets['Qty% Sold Item Shop '+str(i[1])]
    format_1=workbook.add_format({'num_format':'0.00%'})
    format_2=workbook.add_format({'num_format':'#,##0'})
    worksheet.set_column('B:B',30)
    worksheet.set_column('C:CC',12,format_2)
    worksheet_1.set_column('B:B',30)
    worksheet_1.set_column('C:CC',12,format_1)
    worksheet_2.set_column('B:B',30)
    worksheet_2.set_column('C:CC',12,format_2)
    worksheet_3.set_column('B:B',30)
    worksheet_3.set_column('C:CC',12,format_1)
    writer.close()
```
