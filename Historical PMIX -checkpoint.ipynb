```python
data=all_data[all_data.Year==2023]
def category(data, 
         dma=None, coop=None, msa=None, shopid=None, state=None, region=None,
         sv=None, comp=None, week_min=1, week_max=None):
    """
    Generate a pivot table summarizing sales and quantities for different filters.

    Parameters:
        data (pd.DataFrame): Dataset containing sales data.
        dma (list, optional): List of DMAs to filter by.
        coop (list, optional): List of CoOps to filter by.
        msa (str, optional): Substring to match in MSA names.
        shopid (list, optional): List of ShopIDs to filter by.
        state (list, optional): List of states to filter by.
        region (list, optional): List of regions to filter by.
        sv (str, optional): SV filter (default is '').
        comp (str, optional): Comp filter (default is '').
        week_min (int, optional): Minimum week number for filtering (default is 1).
        week_max (int, optional): Maximum week number for filtering (default is data's max week).

    Returns:
        pd.DataFrame: Pivot table summarizing sales and quantities by `ProductTypeName`, 
                      including percentages of sales and quantity.
    """
    # Set default week_max if not provided
    if week_max is None:
        week_max = data.Week.max()

    # Create a base filter
    base_filter = (data.Week.between(week_min, week_max))

    # Add additional filters based on provided arguments
    if shopid:
        base_filter &= data.ParentID.isin(shopid)
    if dma:
        base_filter &= data.DMA.isin(dma)
    if coop:
        base_filter &= data.CoOpName.isin(coop)
    if msa:
        base_filter &= data.MSAName.str.contains(msa, na=False)
    if state:
        base_filter &= data.State.isin(state)
    if region:
        base_filter &= data.Region.isin(region)
    if sv:
        base_filter &= (data.SpecialVenue==sv)
    if comp:
        base_filter &= (data.Comps==comp)

    # Filter the data
    filtered_data = data[base_filter]

    # Check for empty filtered data
    if filtered_data.empty:
        return pd.DataFrame({'Message': ['No data found for the given filters']})

    # Calculate total sales and quantities
    total_sales = filtered_data.SalesMny.sum()
    total_qty = filtered_data.SalesQty.sum()

    # Create the pivot table
    item24 = filtered_data.pivot_table(
        index='ProductTypeName',
        values=['SalesMny', 'SalesQty'],
        aggfunc='sum'
    ).reset_index()

    # Add percentage columns
    item24['Sales%'] = item24['SalesMny'] / total_sales
    item24['Qty%'] = item24['SalesQty'] / total_qty
    item24=item24.rename(columns={'SalesMny':'Sales','SalesQty':'Volume'})
    
    return item24
```

```python
data=all_data[all_data.Year==2023]
def item(data, 
         dma=None, coop=None, msa=None, shopid=None, state=None, region=None,
         sv=None, comp=None, week_min=1, week_max=None):
    """
    Generate a pivot table summarizing sales and quantities for different filters.

    Parameters:
        data (pd.DataFrame): Dataset containing sales data.
        dma (list, optional): List of DMAs to filter by.
        coop (list, optional): List of CoOps to filter by.
        msa (str, optional): Substring to match in MSA names.
        shopid (list, optional): List of ShopIDs to filter by.
        state (list, optional): List of states to filter by.
        region (list, optional): List of regions to filter by.
        sv (str, optional): SV filter (default is '').
        comp (str, optional): Comp filter (default is '').
        week_min (int, optional): Minimum week number for filtering (default is 1).
        week_max (int, optional): Maximum week number for filtering (default is data's max week).

    Returns:
        pd.DataFrame: Pivot table summarizing sales and quantities by `ProductTypeName`, 
                      including percentages of sales and quantity.
    """
    # Set default week_max if not provided
    if week_max is None:
        week_max = data.Week.max()

    # Create a base filter
    base_filter = (data.Week.between(week_min, week_max))

    # Add additional filters based on provided arguments
    if shopid:
        base_filter &= data.ParentID.isin(shopid)
    if dma:
        base_filter &= data.DMA.isin(dma)
    if coop:
        base_filter &= data.CoOpName.isin(coop)
    if msa:
        base_filter &= data.MSAName.str.contains(msa, na=False)
    if state:
        base_filter &= data.State.isin(state)
    if region:
        base_filter &= data.Region.isin(region)
    if sv:
        base_filter &= (data.SpecialVenue==sv)
    if comp:
        base_filter &= (data.Comps==comp)

    # Filter the data
    filtered_data = data[base_filter]

    # Check for empty filtered data
    if filtered_data.empty:
        return pd.DataFrame({'Message': ['No data found for the given filters']})

    # Calculate total sales and quantities
    total_sales = filtered_data.SalesMny.sum()
    total_qty = filtered_data.SalesQty.sum()

    # Create the pivot table
    item23 = filtered_data.pivot_table(
        index='ProductSizeName',
        values=['SalesMny', 'SalesQty'],
        aggfunc='sum'
    ).reset_index()

    # Add percentage columns
    item23['Sales%'] = item23['SalesMny'] / total_sales
    item23['Qty%'] = item23['SalesQty'] / total_qty
    item23=item23.rename(columns={'SalesMny':'Sales','SalesQty':'Volume'})

    return item23
```

```python
def growth(data=all_data,dma=None,coop=None,msa=None,shopid=None,week_min=1,
           week_max=None,state=None,sv=None,region=None,comp=None):
    '''
    Generate a pivot table summarizing sales and quantities YoY Growth for different filters.

    Parameters:
        data (pd.DataFrame): Dataset containing sales data.
        dma (list, optional): List of DMAs to filter by.
        coop (list, optional): List of CoOps to filter by.
        msa (str, optional): Substring to match in MSA names.
        shopid (list, optional): List of ShopIDs to filter by.
        state (list, optional): List of states to filter by.
        region (list, optional): List of regions to filter by.
        sv (list, optional): SV filter (default is '').
        comp (list, optional): Comp filter (default is '').
        week_min (int, optional): Minimum week number for filtering (default is 1).
        week_max (int, optional): Maximum week number for filtering (default is data's max week).

    Returns:
        pd.DataFrame: Pivot table summarizing sales and quantities by Year and, 
                      including percentages of YoY sales and quantity growth.
    
    '''
    
    # Set default week_max if not provided
    if week_max is None:
        week_max = data.Week.max()
    
    # Create a base filter
    base_filter = (data.Week.between(week_min, week_max))
    
    if dma:
        base_filter &= data.DMA.isin(dma)
    if coop:
        base_filter &= data.CoOpName.isin(coop)
    if msa:
        base_filter &= data.MSA.isin(msa)
    if shopid:
        base_filter &= data.ShopID.isin(shopid)
    if state:
        base_filter &= data.State.isin(state)
    if sv:
        base_filter &= (data.SpecialVenue==sv)
    if region:
        base_filter &= data.Region.isin(region)
    if comp:
        base_filter &= (data.Comps==comp)
        
    # Filter the data
    filtered_data = data[base_filter]

    # Check for empty filtered data
    if filtered_data.empty:
        return pd.DataFrame({'Message': ['No data found for the given filters']})
    
    # Calculate total sales and quantities
    total_sales = filtered_data.SalesMny.sum()
    total_qty = filtered_data.SalesQty.sum()
    
    #Defines the column names of the final Pivot Table
    column_names=['ProductSizeName','Sales_23','Sales_24','Units_Sold_23','Units_Sold_24']
    
    #Creates the Pivot Table Based off the condtions that were defined in the function
    yoy_growth=filtered_data.pivot_table(index='ProductSizeName',
                           columns='Year',
                           values=['SalesMny','SalesQty'],
                           aggfunc='sum',fill_value=0).reset_index()
    
    #Renames the columns of the pivot table
    yoy_growth.columns=column_names
    
    #Adds the Growth Columns to the Pivot Table
    yoy_growth['Sales_Growth%']=yoy_growth['Sales_24']/yoy_growth['Sales_23']-1
    yoy_growth['Unit_Growth%']=yoy_growth['Units_Sold_24']/yoy_growth['Units_Sold_23']-1
    
    # Aligns the columns correctlt
    yoy_growth=yoy_growth.iloc[:,[0,2,1,5,4,3,6]]
    
    # Replaces the NaN's, inf, and -inf with 0. This occurs when we try to divide by zero. Math says you can't do that
    yoy_growth=yoy_growth.replace([np.NaN,np.inf,-np.inf],0).sort_values(by='Sales_24',ascending=False)
    
    return(yoy_growth)
```
