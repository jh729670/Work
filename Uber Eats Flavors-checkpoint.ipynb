```python
import pandas as pd
import glob
import zipfile
import os
import pyodbc as odbc
import matplotlib 
import seaborn as sns
import matplotlib.pyplot as plt
import re
import warnings

warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
# Names the needed driver, server name, and database
driver_name='SQL Server'
server_name='10.0.1.127'
database='MAS_RFC'

#Creates the connection string
connection_string=f"""
DRIVER={{{driver_name}}};
SERVER={server_name};
DATABASE={database};
Trust_Connection=yes"""

connection=odbc.connect(connection_string)
print(connection)

# Individual Tables that make up Invoice History
invoice_history=pd.read_sql_query("""SELECT * from MAS_RFC.dbo.AR_InvoiceHistoryHeader Where Year(InvoiceDate)>=2022""",connection)
customer=pd.read_sql_query('''select * from MAS_RFC.dbo.AR_Customer''',connection)
invoice_detail=pd.read_sql_query('''select InvoiceNo,ItemCode,ItemCodeDesc,ItemType,UnitPrice,QuantityShipped, UDF_ROYALTY_AMT,UDF_STD_UNIT_PRICE,ProductLine,CommentText from MAS_RFC.dbo.AR_InvoiceHistoryDetail''',connection)
product_line=pd.read_sql_query('''select * from MAS_RFC.dbo.IM_ProductLine''',connection)
```

```python
# Combines the 4 tables that make up the Invoice History File
inv_hist=invoice_history[['CustomerNo','InvoiceNo','InvoiceType','ShipDate','InvoiceDate','BillToState']] # Only Columns we need
inv_merged=inv_hist.merge(invoice_detail,on='InvoiceNo',how='left') # Merges on InvoiceNo in inv_hist
inv_merge=inv_hist.merge(invoice_detail,on='InvoiceNo',how='inner').drop_duplicates(keep='first')
inv_merge['Ext Base Cost']=inv_merge.QuantityShipped*inv_merge.UDF_STD_UNIT_PRICE # Creates the Product Cost
inv_merge['Ext Royalty']=inv_merge.QuantityShipped*inv_merge.UDF_ROYALTY_AMT # Creates the royalty calc
cust=customer[['CustomerNo','CustomerName','City','ARDivisionNo','UDF_ROYALTY_METHOD','UDF_DISTRIBUTOR']] # Only Columns we need
all_merge=inv_merge.merge(cust,on='CustomerNo',how='inner').drop_duplicates(keep='first') # Merges and drops duplicates rows
filtered=all_merge[(all_merge.ItemType=='1')&(all_merge.ARDivisionNo=='01')]
invoice=filtered.merge(product_line[['ProductLine','ProductLineDesc']],on='ProductLine').drop_duplicates(keep='first') # merges and drops duplicate rows
```

```python
# Gets the Unique Flavors from Invoice History
mixes=['CUSTARD','MIXES','CUSTARD - KOSHER','MIXES-SUGARFREE']
invoice['ItemCodeDesc']=invoice.ItemCodeDesc.str.split('(').str[0] # Separtes out the text in the Item Code Desc
flavors=invoice[invoice.ProductLineDesc.isin(mixes)] # Gets only the flavors in the Item Code Desc
flavors.ItemCodeDesc=flavors.ItemCodeDesc.str.title() # Upper Cases the First Letter only
flavors.ItemCodeDesc=flavors.ItemCodeDesc.str.replace('Choco Custard','Chocolate Custard') #Changes to Chocolate Custard
flavors.ItemCodeDesc=flavors.ItemCodeDesc.str.replace('Mix','')# Removes Mix
flavors.ItemCodeDesc=flavors.ItemCodeDesc.str.strip()# Removes White Space
```

```python
#Sets up the filter thourgh orders data to find flavors through the various columns in orders dataframe
all_flavors=all_flavors.Flavors.str.lower().str.strip()

# Empty list to store the results
results = []

# Iterate over the flavor columns
for column in ['Flavor_1', 'Flavor_2', 'Flavor_3']:  # List all flavor columns to check
    
    orders[column] = orders[column].str.lower().str.strip()
    # Filter rows where the column contains a flavor in the flavors_list
    filtered_df = orders[orders[column].isin(all_flavors)]
    
    # Append the results: Keep only the order_id and the flavor found
    for idx, row in filtered_df.iterrows():
        results.append({
            'order_id': row['workflow_uuid'],
            'sku_name':row['sku_name'],
            'flavor': row[column]  # This will capture the flavor from the current column
        })

# Convert the results to a DataFrame
df_flavor_orders = pd.DataFrame(results)

#print(df_flavor_orders)
```

```python


# Pattern to match and prioritize "vanilla custard" or "chocolate custard"
pattern = re.compile(r'(vanilla custard|chocolate custard|small|large)', re.IGNORECASE)

# Function to reorder terms with "large" or "small" always as the first word
def prioritize_terms(text):
    # Find all prioritized matches
    matches = pattern.findall(text)
    # Separate "large" or "small" if present
    size = [word for word in matches if word.lower() in ["large", "small"]]
    other_matches = [word for word in matches if word.lower() not in ["large", "small"]]
    
    # Check if "gelati" is in the text
    if "gelati" in text.lower():
        remaining_text = pattern.sub('', text).replace("gelati", "").strip()
        return ' '.join(size + other_matches + [remaining_text, "gelati"]).strip()
    else:
        remaining_text = pattern.sub('', text).strip()
        return ' '.join(size + other_matches + [remaining_text]).strip()

```
