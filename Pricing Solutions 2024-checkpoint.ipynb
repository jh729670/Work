```python
#Imports the necessary libraries
import pandas as pd
import os
import numpy as np
import pyodbc as odbc
from datetime import date,timedelta,datetime
import datetime as dt
import xlsxwriter
os.chdir(r'L:\J.Harned\Calendar')
calendar=pd.read_excel('2024 Sales Calendar.xlsx')
calendar['Year']=calendar.StartDate.dt.year
os.chdir(r'G:\FinanceReports\2024\Wk41')
# Names the needed driver, server name, and database
driver_name='SQL Server'
server_name='10.0.1.218'
database='ritascoolnet_live'
#Creates the connection string
connection_string=f"""
DRIVER={{{driver_name}}};
SERVER={server_name};
DATABASE={database};
Trust_Connection=yes"""
#Creates the Connection to SQL Server
connection=odbc.connect(connection_string)
print(connection)
```

```python
# Queries the tables from SQL
# Reads in the Weekly Sales in 2023
data=pd.read_sql_query('''SELECT *
  FROM [ritas_bi].[dbo].[tmp_raw_plda]
  WHERE YEAR(StartDate)>=2023
  ORDER BY Week DESC''',connection)
#ShopDetail
data3=pd.read_sql_query('''Select intstoreid as intStoreID,RitasID as ShopID, FC,CoolNetGroup,LegacyStoreType,State,MSAName,DMA,StoreName as ShopName,
CoOpName,Franchisee,[Parent Shop #] as ParentID,[Special Venue] as SV,StoreStatus,[Re Open Date] as ReOpenDate,OpenDate,StoreType,Seasonality from ritascoolnet_live.dbo.vShopDetail''',connection)
relo=pd.read_sql_query('''Select * from ritascoolnet_live.dbo.GOOGLE_MAP_store''',connection)
```

```python
retail=data[data.StartDate>='2024-01-01'].groupby(['ShopID','ProdSizeCode'])[['SalesMny','SalesQty']].sum().reset_index()
retail['Retail_Price']=retail.SalesMny.div(retail.SalesQty)
retail_price=retail.fillna(0)
```

```python
retail_price
```

```python
store_list=psl_data[psl_data.StartDate>='2024-01-01'][['intStoreID','ShopID','ShopName','Region','State','MSAName','ShopStatus','StreetAddress1','City','ZipCode',
                    'FC','Seasonality','SeasonalOpenDate','SeasonalCloseDate','LegacyShopType','LegacyShopStatus',
                    'SpecialVenue','Latitude','Longitude']]
store_list=store_list.sort_values(by=['intStoreID','ShopID']).drop_duplicates(keep='first',subset='ShopID')
```

```python
store_list
```

```python
os.chdir(r'G:\FinanceReports\2024\PSL Data')
with pd.ExcelWriter('2024 Store List.xlsx',datetime_format='mm/dd/yyyy') as writer1:
    store_list.sort_values(by='ShopID').to_excel(writer1, sheet_name='StoreInfo',startrow=5,startcol=0,index=False)
    
    wb=writer1.book
    word_format=wb.add_format({'align':'center','valign':'center'})
    border_format=wb.add_format({'right':1,'left':1,'bottom':1,'top':1})
    header_format=wb.add_format({'bold':True,
                                  'italic':True})
    ws=writer1.sheets['StoreInfo']
    
    ws.conditional_format(xlsxwriter.utility.xl_range(5,0,len(store_list)+5,len(store_list.columns)-1),
                              {'type': 'no_errors', 'format': border_format})
    
    text1='Ritas Franchise Operations,LLC'
    text2='2024 Ritas Store Info'
    text3='Pricing Solutions'
    text4='Prepapred on '+str(date.today().strftime('%m-%d-%y'))
    
    ws.write(0,1,text1,header_format)
    ws.write(1,1,text2,header_format)
    ws.write(2,1,text3,header_format)
    ws.write(3,1,text4,header_format)
    
    ws.set_column('A:Z',16,word_format)
    ws.set_column('B:B',40,word_format)
    ws.set_column('C:C',40,word_format)
    ws.set_column('F:F',40,word_format)
    ws.set_column('H:H',40,word_format)
    ws.set_column('K:K',26,word_format)
    ws.set_column('R:S',20,word_format)
    ws.set_column('M:N',20,word_format)
    ws.hide_gridlines(2)
    ws.insert_image('A1','Logo.png')
```

```python
with pd.ExcelWriter('2024 Retail Pricing by Shop.xlsx',datetime_format='mm/dd/yyyy') as writer1:
    retail_price.sort_values(by='ShopID').to_excel(writer1, sheet_name='RetailPrice',startrow=5,startcol=0,index=False)
    
    wb=writer1.book
    word_format=wb.add_format({'align':'center','valign':'center'})
    border_format=wb.add_format({'right':1,'left':1,'bottom':1,'top':1})
    header_format=wb.add_format({'bold':True,
                                  'italic':True})
    dollar_format=wb.add_format({'num_format':'$#,##0.00',
                                'align':'center',
                                'valign':'vcenter'})
    qty_format=wb.add_format({'num_format':'#,##0',
                                'align':'center',
                                'valign':'vcenter'})
    
    ws=writer1.sheets['RetailPrice']
    
    ws.conditional_format(xlsxwriter.utility.xl_range(5,0,len(retail_price)+5,len(retail_price.columns)-1),
                              {'type': 'no_errors', 'format': border_format})
    
    text1='Ritas Franchise Operations,LLC'
    text2='2024 Retail Pricing by Shop'
    text3='Pricing Solutions'
    text4='Prepapred on '+str(date.today().strftime('%m-%d-%y'))
    
    ws.write(0,1,text1,header_format)
    ws.write(1,1,text2,header_format)
    ws.write(2,1,text3,header_format)
    ws.write(3,1,text4,header_format)
    
    ws.set_column('A:B',12,word_format)
    ws.set_column('C:C',16,dollar_format)
    ws.set_column('D:D',16,qty_format)
    ws.set_column('E:E',16,dollar_format)
    ws.hide_gridlines(2)
    ws.insert_image('A1','Logo.png')
```

```python
with pd.ExcelWriter('WeeklySales Oct 2023-Sept 2024.xlsx',datetime_format='mm/dd/yyyy') as writer1:
    psl_data.sort_values(by='ShopID').to_excel(writer1, sheet_name='Data',startrow=5,startcol=0,index=False)
    
    wb=writer1.book
    word_format=wb.add_format({'align':'center','valign':'center'})
    border_format=wb.add_format({'right':1,'left':1,'bottom':1,'top':1})
    header_format=wb.add_format({'bold':True,
                                  'italic':True})
    dollar_format=wb.add_format({'num_format':'$#,##0.00',
                                'align':'center',
                                'valign':'vcenter'})
    qty_format=wb.add_format({'num_format':'#,##0',
                                'align':'center',
                                'valign':'vcenter'})
    
    ws=writer1.sheets['Data']
    
    ws.conditional_format(xlsxwriter.utility.xl_range(5,0,len(psl_data)+5,len(psl_data.columns)-1),
                              {'type': 'no_errors', 'format': border_format})
    
    text1='Ritas Franchise Operations,LLC'
    text2='Weekly Sales Data from Oct 2023-Sept 2024'
    text3='Pricing Solutions'
    text4='Prepapred on '+str(date.today().strftime('%m-%d-%y'))
    
    ws.write(0,1,text1,header_format)
    ws.write(1,1,text2,header_format)
    ws.write(2,1,text3,header_format)
    ws.write(3,1,text4,header_format)
    
    ws.set_column('A:AZ',16,word_format)
    ws.set_column('H:H',16,dollar_format)
    ws.set_column('G:G',16,qty_format)
    ws.set_column('N:N',40,word_format)
    ws.set_column('W:X',26,word_format)
    ws.set_column('AC:AD',40, word_format)
    ws.set_column('AF:AG',20,word_format)
    ws.set_column('AK:AL',40,word_format)
    ws.hide_gridlines(2)
    ws.insert_image('A1','Logo.png')
```
