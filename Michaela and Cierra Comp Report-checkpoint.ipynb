```python
import pandas as pd
import os
import numpy as np
import pyodbc as odbc
from datetime import date,timedelta
import xlsxwriter
os.chdir(r'L:\J.Harned\Calendar')
calendar=pd.read_excel('2024 Sales Calendar.xlsx')
calendar['Year']=calendar.StartDate.dt.year
os.chdir(r'G:\FinanceReports\2024\Monthly Comps')
#Names the needed driver, server name, and database
driver_name='SQL Server'
server_name='10.0.1.218'
database='ritascoolnet_live'
#Creates the connection string
connection_string=f"""
DRIVER={{{driver_name}}};
SERVER={server_name};
DATABASE={database};
Trust_Connection=yes"""
#Creates the Connection to SQL Server
connection=odbc.connect(connection_string)
print(connection)
# Queries the tables from SQL
# Reads in the Weekly Sales in 2023
data=pd.read_sql_query('''SELECT *
  FROM [ritas_bi].[dbo].[tmp_raw_plda]
  WHERE YEAR(StartDate)=2024
  ORDER BY Week DESC''',connection)
# Reads in the Weekly Sales in 2022
df=pd.read_sql_query('''SELECT *
  FROM [ritas_bi].[dbo].[tmp_raw_plda]
  WHERE YEAR(StartDate)=2023 AND Week <=40''',connection)
data3=pd.read_sql_query('''Select intstoreid,RitasID as ShopID, FC,CoolNetGroup,LegacyStoreType,State,MSAName,DMA,StoreName as ShopName,StoreStatus,CoOpName,[Special Venue] as SV,[Parent Shop #] as StringID from ritascoolnet_live.dbo.vShopDetail''',connection)
product_sales=pd.read_sql_query('''Select * from ritascoolnet_live.dbo.tblProductSalesTransactions Where Year(dteStartDate)>=2023 ''',connection)
ops=pd.read_sql_query('''select intstoreid as intStoreID, RitasID as ShopID,FSMName as FC,OperationsDirector as Director from ritas_bi.dbo.shopSalesRankingWeekly''',connection)
```

```python
df.loc[df.StartDate=='2023-03-20','Week']=12
df.loc[df.StartDate=='2023-03-27','Week']=13
```

```python
total=pd.concat([data,df])
weeks=total[['StartDate','Week']].drop_duplicates(keep='first')
weeks
```

```python
product_sales=product_sales.rename(columns={'dteStartDate':'StartDate'})
product_sales=product_sales.merge(weeks,on='StartDate',how='left')
product_sales['Retail_GC']=product_sales.intTransactionCount-product_sales.intUECount\
-product_sales.intDDCount-product_sales.intGHCount
product_sales['Year']=product_sales.StartDate.dt.year
product_sales=product_sales.rename(columns={'intRitasStoreID':'ShopID'})
product_sales.ShopID=product_sales.ShopID.astype('int')
product_sales=product_sales[product_sales.Week<=data.Week.max()]
```

```python
# Pulls Marketing info needed like Coolnet Group, MSA, and DMA
store=pd.read_sql_query('''Select intstoreid as intStoreID,RitasID as ShopID, FC,CoolNetGroup,LegacyStoreType,State,MSAName,DMA,StoreName as ShopName,CoOpName,OpenDate,StoreStatus,[Re Open Date] as ReOpenDate from ritascoolnet_live.dbo.vShopDetail''',
                        connection)
# relo=pd.read_sql_query('''Select intStoreID,dteClosedForRelocation,intParentStoreID as ParentID,
# intRitasStoreID as CustomerNo,bitDriveThru as DT from ritascoolnet_live.dbo.GOOGLE_MAP_store''',connection)
# store=store.merge(relo[['intStoreID','dteClosedForRelocation']],on='intStoreID')
# store=store.sort_values(by='intStoreID',ascending=False).drop_duplicates(subset='RitasID',keep='first')
# store=store.rename(columns={'RitasID':'ShopID'})
# store['dteClosedForRelocation']=pd.to_datetime(store['dteClosedForRelocation'])
# store['Days']=(store.ReOpenDate-store.dteClosedForRelocation).dt.days
# store['OpenDate']=np.where(store.Days>=365,store.ReOpenDate,store.OpenDate)
# # open_shops=store[store.StoreStatus=='Open']
```

```python
test=total.merge(store[['intStoreID','ShopID','StoreStatus','OpenDate']],on=['ShopID','intStoreID'],how='left')
open_sales=test[(test.SpecialVenue=='No')&(test.ShopID<10000)&(test.StoreStatus.str.contains('Open|Transferred'))&(test.OpenDate<='2023-03-05')]
open_sales['Year']=open_sales.StartDate.dt.year
```

```python
num_weeks=open_sales.pivot_table(index='ShopID',columns='Year',values='Week',aggfunc='nunique',fill_value=0).reset_index()
```

```python
num_weeks.rename(columns={2023:'Weeks Open 2023',2024:'Weeks Open 2024'},inplace=True)
```

```python
sales_week=open_sales.pivot_table(index='ShopID',columns='Year',values='SalesMny',aggfunc='sum',fill_value=0).reset_index()
```

```python
sales_week=sales_week[(sales_week[2023]>0)&(sales_week[2024])]
```

```python
sales_week=sales_week.merge(ops[['ShopID','Director','FC']],on='ShopID',how='left')
sales_week=sales_week[['ShopID','Director','FC',2024,2023]]
```

```python
comp_volume=open_sales.pivot_table(index='ShopID',
                                                   columns='Year',
                                                   values='SalesQty',
                                                   aggfunc='sum').reset_index()
comp_volume
```

```python
data=data[['intStoreID','ShopID','ShopName']].sort_values(by='ShopID').drop_duplicates(subset='ShopName',keep='last')
```

```python

comp_vol=comp_volume.merge(data[['ShopID','ShopName']],on='ShopID').drop_duplicates(subset=[2024,2023],keep='first')
# comp_vol=comp_vol.drop('StringID',axis=1)
comp_vol=comp_vol[['ShopID','ShopName',2024,2023]]
comp_vol['Qty_Change %']=comp_vol[2024]/comp_vol[2023]-1
comp_vol.rename(columns={2024:'Qty 24',2023:'Qty 23'},inplace=True)
comp_vol
```

```python
comp_s=sales_week.merge(data[['ShopID','ShopName']],on='ShopID').drop_duplicates(subset=[2024,2023],keep='first')
comp_s=comp_s[['ShopID','ShopName','Director','FC',2024,2023]]
comp_s['Sales_Change %']=comp_s[2024]/comp_s[2023]-1
comp_s.rename(columns={2024:'Sales 24',2023:'Sales 23'},inplace=True)
comp_s
```

```python
comp_s[comp_s.ShopID.duplicated()]
```

```python
comp_s.merge(comp_vol,on=['ShopID','ShopName'])
```

```python
gc_piv=product_sales[product_sales.ShopID.isin(comp_s.ShopID)].pivot_table(index='ShopID',
                                                                           columns='Year',
                                                                           values='Retail_GC',aggfunc='sum',fill_value=0).reset_index()
gc_piv['GC%_Chng']=gc_piv[2024]/gc_piv[2023]-1
gc_piv=gc_piv.replace([np.inf],0)
gc_piv=gc_piv[['ShopID',2024,2023,'GC%_Chng']]
gc_piv.rename(columns={2024:'GC 24',2023:'GC 23'},inplace=True)
gc_piv
```

```python
all_comps=comp_s.merge(comp_vol,on=['ShopID','ShopName']).merge(gc_piv,on='ShopID').merge(num_weeks,on='ShopID')
```

```python
all_comps
```

```python
summary=pd.DataFrame({'Statistics':['Sales 23','Sales 24','Qty 23','Qty 24','GC 23','GC 24','Sales Growth','Qty Growth','GC Growth','Weeks Open 2023','Weeks Open 2024'],
                    'Averages':[all_comps['Sales 23'].mean(),all_comps['Sales 24'].mean(),
                                all_comps['Qty 23'].mean(),all_comps['Qty 24'].mean(),all_comps['GC 23'].mean(),all_comps['GC 24'].mean(),all_comps['Sales 24'].sum()/all_comps['Sales 23'].sum()-1,
                      all_comps['Qty 24'].sum()/all_comps['Qty 23'].sum()-1,
                    all_comps['GC 24'].sum()/all_comps['GC 23'].sum()-1,
                               all_comps['Weeks Open 2023'].mean(),
                               all_comps['Weeks Open 2024'].mean()] })
```

```python
timestamp=date.today().strftime('%m-%d-%y')
with pd.ExcelWriter('24 v 23 Retail Comp with Weeks Open thru Wk '+str(total.Week.max())+' '+str(timestamp)+'.xlsx') as writer:
    summary.to_excel(writer,sheet_name='Summary',index=False,startrow=5,startcol=0)
    
    text='Ritas Franchise Company,LLC'
    text1='Summary Statistics Wk '+str(total.Week.max())
    text2='Retail Only (Special Venues, Mobiles, Satellites Excluded)'
    text3='As of '+str(total.EndDate.max().strftime('%m-%d-%y'))
    
    wb=writer.book
    ws=writer.sheets['Summary']
    
    format_header2=wb.add_format()
    format_header2.set_italic()
    format_header2.set_bold()
    border_format=wb.add_format({'top':2,'bottom':2,'left':2,'right':2})
    total_dollar_format=wb.add_format({'num_format':'$#,##0',
                                        'align':'center',
                                        'valign':'vcenter'})
    format_qty=wb.add_format({'num_format':'#,##0',
                                        'align':'center',
                                        'valign':'vcenter'})
    perc_format=wb.add_format({'num_format':'#0.0%',
                                         'align':'center',
                                         'valign':'vcenter'})
    week_format=wb.add_format({'num_format':'#,##0.0',
                                        'align':'center',
                                        'valign':'vcenter'})
    ws.conditional_format(xlsxwriter.utility.xl_range(5,0,len(summary)+5,len(summary.columns)-1),
                                  {'type': 'no_errors', 'format': border_format})
    word_format=wb.add_format({'align':'center','valign':'vcenter'})
    
    ws.write(0,1,text,format_header2)
    ws.write(1,1,text1,format_header2)
    ws.write(2,1,text2,format_header2)
    ws.write(3,1,text3,format_header2)
    ws.set_row(6,None,total_dollar_format)
    ws.set_row(7,None,total_dollar_format)
    ws.set_row(8,None,format_qty)
    ws.set_row(9,None,format_qty)
    ws.set_row(10,None,format_qty)
    ws.set_row(11,None,format_qty)
    ws.set_row(12,None,perc_format)
    ws.set_row(13,None,perc_format)
    ws.set_row(14,None,perc_format)
    ws.set_row(15,None,week_format)
    ws.set_row(16,None,week_format)
    ws.set_column('A:A',17,None)
    ws.hide_gridlines(2)
    ws.insert_image('A1','Logo.png')
    
    all_comps.to_excel(writer, sheet_name='Comp Shops',index=False,startrow=5,startcol=0)
    ws1=writer.sheets['Comp Shops']
    ws1.conditional_format(xlsxwriter.utility.xl_range(5,0,len(all_comps)+5,len(all_comps.columns)-1),
                                  {'type': 'no_errors', 'format': border_format})
    
    ws1.write(0,1,text,format_header2)
    ws1.write(1,1,text1,format_header2)
    ws1.write(2,1,text2,format_header2)
    ws1.write(3,1,text3,format_header2)
    ws1.set_column('A:A',10,word_format)
    ws1.set_column('B:D',45,word_format)
    ws1.set_column('E:F',16,total_dollar_format)
    ws1.set_column('G:G',16,perc_format)
    ws1.set_column('H:I',16,format_qty)
    ws1.set_column('J:J',16,perc_format)
    ws1.set_column('K:L',16,format_qty)
    ws1.set_column('M:M',16,perc_format)
    ws1.set_column('N:O',16,word_format)
    ws1.hide_gridlines(2)
    ws1.insert_image('A1','Logo.png')
```

```python
week_sales=open_sales[(open_sales.ShopID<10000)&(open_sales.SpecialVenue=='No')&(open_sales.Year==2024)].pivot_table(index=['ShopID'],columns=['EndDate'],values='SalesMny',aggfunc='sum',fill_value=0)
weeks_open=open_sales[(open_sales.ShopID<10000)&(open_sales.SpecialVenue=='No')&(open_sales.Year==2024)].pivot_table(index=['ShopID'],columns=['EndDate'],values='Week',aggfunc='nunique',fill_value=0,
                                                                                     margins=True)
combined=week_sales.merge(weeks_open.iloc[:,-1],left_index=True,right_index=True)
combined=combined.rename(columns={'All':'Weeks Open'})
with pd.ExcelWriter('Michaela Weeks Open 2024 Shops Wk '+str(total.Week.max())+' '+str(timestamp)+'.xlsx',datetime_format='mm/dd/yyyy') as writer:
    
    combined.to_excel(
        writer, engine='xlsxwriter',sheet_name='2024 Weeks Open',startrow=5,startcol=0)
    wb=writer.book
    ws=writer.sheets['2024 Weeks Open']
    text='Ritas Franchise Company,LLC'
    text1='2024 Sales and Weeks Open thru '+str(total.Week.max())
    text2='Retail Only (Special Venues, Mobiles, Satellites Excluded)'
    text3='As of '+str(total.EndDate.max().strftime('%m-%d-%y'))
    total_dollar_format=wb.add_format({
                                          'num_format':'$#,##0',
                                        'align':'center',
                                        'valign':'vcenter'})
    format_qty=wb.add_format({'num_format':'#,##0',
                                        'align':'center',
                                        'valign':'vcenter'})
    border_format=wb.add_format({'top':2,'bottom':2,'left':2,'right':2})
    a1=wb.add_format({'align':'center','valign':'vcenter','bold':True})
    format_header2=wb.add_format()
    format_header2.set_italic()
    format_header2.set_bold()
    ws.conditional_format(xlsxwriter.utility.xl_range(5,0,len(combined)+5,len(combined.columns)),
                                  {'type': 'no_errors', 'format': border_format})
    
    ws.write(0,1,text,format_header2)
    ws.write(1,1,text1,format_header2)
    ws.write(2,1,text2,format_header2)
    ws.write(3,1,text3,format_header2)
    
    ws.set_column('A:A',12,a1)
    ws.set_column('B:CZ',12,total_dollar_format)
    ws.set_column('AQ:AQ',12,format_qty) #Change this every week
    ws.hide_gridlines(2)
    ws.insert_image('A1','Logo.png')
```
