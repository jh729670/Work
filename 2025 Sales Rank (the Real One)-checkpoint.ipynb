```python
def get_initials(name):
    if not isinstance(name, str) or name.strip() == "":  # Handle None or empty strings
        return name  # Keep None or empty as is
    parts = name.split()
    if len(parts) == 1:  # If only one name, return it as is
        return name
    else:
        return ''.join([part[0].upper() for part in parts])  # Extract initials

data3['FC Initials'] = data3['FC'].apply(get_initials)
data3.loc[data3.FC=='Joe Toocheck','FC Initials']='JToo'
data3.loc[data3.FC=='John Martin (FC)','FC Initials']='JM'
```

```python
#State Growth
core_sales=sales[(sales.Year>=2024)&(sales.State.isin(core))&(sales.Week<=data.Week.max())].pivot_table(
index='State',columns='Year',values=['SalesMny','ParentID'],aggfunc={'SalesMny':'sum','ParentID':'nunique'}).reset_index()
core_sales['Growth']=core_sales.iloc[:,-1].div(core_sales.iloc[:,-2])-1
dev_sales=sales[(sales.Year>=2024)&~(sales.State.isin(core))&(sales.Week<=data.Week.max())].pivot_table(
index='State',columns='Year',values=['SalesMny','ParentID'],aggfunc={'SalesMny':'sum','ParentID':'nunique'},fill_value=0).reset_index()
dev_sales['Growth']=dev_sales.iloc[:,-1].div(dev_sales.iloc[:,-2])-1
dev_sales.Growth.replace([np.inf,np.nan],0,inplace=True)
```

```python
#Comp Store Analysis
comp=cy_pivot[(cy_pivot.loc[:,'Sales_25']>0)&(cy_pivot.loc[:,'Sales_24']>0)]
core_comp=comp[comp.State.isin(core)]
develop_comp=comp[~comp.State.isin(core)]
core_dt_comp=core_comp[core_comp.DT=='Yes']
develop_dt_comp=develop_comp[develop_comp.DT=='Yes']
core_comp_shops=core_comp.ParentID
develop_comp_shops=develop_comp.ParentID
```

```python
#3rd Party Sales State Analysis
core_all=core_3pd.pivot_table(index='State',columns='Year',values='SalesMny',aggfunc='sum').reset_index()
core_all['Growth']=core_all[2025]/core_all[2024]-1
core_all=core_all[['State',2025,2024,'Growth']]
dev_all=dev_3pd.pivot_table(index='State',columns='Year',values='SalesMny',aggfunc='sum',fill_value=0).reset_index()
dev_all['Growth']=dev_all[2025]/dev_all[2024]-1
dev_all=dev_all[['State',2025,2024,'Growth']]
dev_all.replace([np.inf,np.nan],0,inplace=True)
#3rd Party Comp State Analysis
core_comp_3pd=core_3pd[core_3pd.ParentID.isin(core_comp_shops)].pivot_table(
    index='State',columns='Year',values='SalesMny',aggfunc='sum').reset_index()
core_comp_3pd['Growth']=core_comp_3pd[2025]/core_comp_3pd[2024]-1
core_comp_3pd=core_comp_3pd[['State',2025,2024,'Growth']]
dev_comp_3pd=dev_3pd[dev_3pd.ParentID.isin(develop_comp_shops)].pivot_table(index='State',columns='Year',
                                                                     values='SalesMny',aggfunc='sum',fill_value=0).reset_index()
dev_comp_3pd['Growth']=dev_comp_3pd[2025]/dev_comp_3pd[2024]-1
dev_comp_3pd=dev_comp_3pd[['State',2025,2024,'Growth']]
dev_comp_3pd.Growth.replace([np.inf,np.nan],0,inplace=True)
```
