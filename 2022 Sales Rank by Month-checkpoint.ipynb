```python
data[(data.Week.isin(Jan))&(data.YTDTotalShopSales!=0)&(~data.RitasID.isin(remove))].groupby(['RitasID','StoreName','State','SpecialVenue','SiteType'])[['RetailSalesYTD',
                                                                                                         'MobileSalesYTD',
                                                                                                         'CateringSales',
                                                                                                         'DigitalSalesYTD',
                                                                                                         'SatelliteSalesYTD',
                                                                                                         'YTDTotalShopSales']].agg('sum').reset_index().sort_values(by='YTDTotalShopSales',ascending=False)
```
