```python
# Names the needed driver, server name, and database
driver_name='SQL Server'
server_name='10.0.1.218'
database='ritascoolnet_live'
```

```python
#Creates the connection string
connection_string=f"""
DRIVER={{{driver_name}}};
SERVER={server_name};
DATABASE={database};
Trust_Connection=yes"""
```

```python
#Creates the Connection to SQL Server
connection=odbc.connect(connection_string)
print(connection)
```

```python
# Queries the tables from SQL
# Reads in the Daily Sales from 2023
data1=pd.read_sql_query('''SELECT * FROM [dbo].[smartDailySales] WHERE YEAR(dteDate)=2023''',connection)
# Reads in the Google Map to be able to join the inStoreID to the RitasID
data2=pd.read_sql_query('''Select intStoreID, intRitasStoreID, vchStoreState,vchStoreName from ritascoolnet_live.dbo.GOOGLE_MAP_store''',connection)
# Reads in the Daily Guest Count from 2023
data3=pd.read_sql_query('''SELECT * from ritascoolnet_live.dbo.smartDailyZTape Where intZTapeID=6 and Year(dteDate)=2023''',connection)
```

```python
#Ceates the start and end date to veiw the data
start=data1.dteDate.max()-timedelta(days=27)
end=data1.dteDate.max()
```

```python
start
```

```python
#Filters Data on the start and end date
data1=data1[(data1.dteDate>=start)&(data1.dteDate<=end)]
```

```python
#Filters Guest Count on the Start and End Date
data3=data3[(data3.dteDate>=start)&(data3.dteDate<=end)]
#Changes the Guest Count from a string to an integer
data3.vchValue=data3.vchValue.astype('int')
```

```python
# Merges the three the Dataframes into one
df=pd.merge(data1,data2,on='intStoreID')
```

```python
#Groups sum of Guest Count by Store to get total guest count by date range
gc=data3.groupby('intStoreID')['vchValue'].sum().reset_index()
```

```python
gc
```

```python
# Sums up all of the sales,Guest Countt and QTY by Shop
allsales=round(df.groupby(['intStoreID','intRitasStoreID','vchStoreName','vchStoreState'])[['intQuantity','mnyPosSales']].sum().reset_index(),2)
allsales.rename(columns={'intQuantity':'QTY','mnyPosSales':'Sales'},inplace=True)
```

```python
#Merges Guest count with QTY and Sales 
allsales=pd.merge(allsales,gc,on='intStoreID')
allsales.rename(columns={'vchValue':'Guest_Count'},inplace=True)
```

```python
allsales.head()
```

```python
# Sums up the the Coffee sales and QTY by size
#coffee=round(df[df.vchProductTypeName=='Frozen Coffee'].groupby(['intStoreID','intRitasStoreID','vchStoreName','vchStoreState','vchProductSizeName'])[['intQuantity','mnyPosSales']].sum().reset_index(),2)
pivot=round(df[(df.vchProductTypeName=='Frozen Coffee')&(df.mnyPrice>0.01)].pivot_table(index=['intStoreID','intRitasStoreID','vchStoreName','vchStoreState'],columns=['vchProductSizeName'],values='intQuantity',aggfunc='sum',fill_value=0),2).reset_index()
#pivot.drop(columns='Regular Frozen Coffee',inplace=True)
pivot.rename(columns={'Large Frozen Coffee':'Qty Large Frozen Coffee','Small Frozen Coffee':'Qty Small Frozen Coffee'},inplace=True)
coffee=round(df[(df.vchProductTypeName=='Frozen Coffee')&(df.mnyPrice>0.01)].pivot_table(index=['intStoreID','intRitasStoreID','vchStoreName','vchStoreState'],columns=['vchProductSizeName'],values='mnyPosSales',aggfunc='sum',fill_value=0),2).reset_index()
#coffee.drop(columns='Regular Frozen Coffee',inplace=True)
coffee.rename(columns={'Large Frozen Coffee':'Sales Large Frozen Coffee','Small Frozen Coffee':'Sales Small Frozen Coffee'},inplace=True)
```

```python
discount=df[(df.mnyPrice<=0.01)&(df.vchProductTypeName=='Frozen Coffee')]
discount=round(discount.pivot_table(index=['intStoreID','intRitasStoreID','vchStoreName','vchStoreState'],columns=['vchProductSizeName'],values='intQuantity',aggfunc='sum',fill_value=0),2).reset_index()
discount.drop(columns='Regular Frozen Coffee',inplace=True)
discount.rename(columns={'Large Frozen Coffee':'# Discount Large Frozen Coffee','Small Frozen Coffee':'# Discount Small Frozen Coffee'},inplace=True)
```

```python
# Merges the all sales and qty with sales and QTY of Frozen Coffee
total=pd.merge(allsales,pivot,how='outer').merge(coffee,how='outer').merge(discount,how='outer').fillna(0)
total.intRitasStoreID=total.intRitasStoreID.astype('int')
#total.rename(columns={'intQuantity':'Coffee Qty','mnyPosSales':'Coffee Sales'},inplace=True)
# Formulates the amount of coffee sold for every 1000 transactions
total['Coffee_per_1000']=round((total['Qty Large Frozen Coffee']+total['Qty Small Frozen Coffee'])/total.QTY*1000,2)
```

```python
# Sorts the dataframe by RitasID and sends to excel
total.sort_values(by='intRitasStoreID').to_excel('Coffee Sales 5-15-2023.xlsx',sheet_name='4.17-5.14',index=False)
```
