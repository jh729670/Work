```python
# Creates function that will allow user to enter in parameters to view sales and volume growth for 2024 vs 2023.
# Users can view this by shop, DMA, CoOp, state or region
# Users can also choose to veiw data by timeframe and whether they want Special Venues to be included in the analysis
# or whether the user wants to veiw Comp shops only

def growth(dma=[],coop=[],msa='',shopid=[],week_min=1,week_max=all_data.Week.max(),state=[],sv='',region=[],comp=''):
    column_names=['ProductTypeName','Sales_23','Sales_24','Units_Sold_23','Units_Sold_24']
    if shopid:
        item23=all_data[(all_data.ShopID.isin(shopid))&(all_data.Week.between(week_min,week_max))&(all_data.SV==sv)&(all_data.Comps==comp)].pivot_table(index='ProductTypeName',columns='Year',
                                               values=['SalesMny','SalesQty'],aggfunc='sum',fill_value=0).reset_index()
        item23.columns=column_names
        item23=item23.iloc[:,[0,2,1,4,3]]
        item23['Sales_Growth%']=item23['Sales_24']/item23['Sales_23']-1
        item23['Unit_Growth%']=item23['Units_Sold_24']/item23['Units_Sold_23']-1
        item23=item23.iloc[:,[0,1,2,5,3,4,6]]
        item23.replace([np.inf,-np.inf],inplace=True)
        item23.fillna(0,inplace=True)
    elif shopid==0:
        if dma:
            item23=all_data[(all_data.DMA.isin(dma))&(all_data.Week.between(week_min,week_max))&(all_data.SV==sv)&(all_data.Comps==comp)].pivot_table(index='ProductTypeName',columns='Year',
                                                   values=['SalesMny','SalesQty'],aggfunc='sum',fill_value=0).reset_index()
            item23.columns=column_names
            item23=item23.iloc[:,[0,2,1,4,3]]
            item23['Sales_Growth%']=item23['Sales_24']/item23['Sales_23']-1
            item23['Unit_Growth%']=item23['Units_Sold_24']/item23['Units_Sold_23']-1
            item23=item23.iloc[:,[0,1,2,5,3,4,6]]
            item23.replace([np.inf,-np.inf],inplace=True)
            item23.fillna(0,inplace=True)
        if coop:
            item23=all_data[(all_data.CoOpName.isin(coop))&(all_data.Week.between(week_min,week_max))&(all_data.SV==sv)&(all_data.Comps==comp)].pivot_table(index='ProductTypeName',columns='Year',
                                                   values=['SalesMny','SalesQty'],aggfunc='sum',fill_value=0).reset_index()
            item23.columns=column_names
            item23=item23.iloc[:,[0,2,1,4,3]]
            item23['Sales_Growth%']=item23['Sales_24']/item23['Sales_23']-1
            item23['Unit_Growth%']=item23['Units_Sold_24']/item23['Units_Sold_23']-1
            item23=item23.iloc[:,[0,1,2,5,3,4,6]]
            item23.replace([np.inf,-np.inf],inplace=True)
            item23.fillna(0,inplace=True)
        if msa:
            item23=all_data[(all_data.MSAName.str.contains(msa))&(all_data.Week.between(week_min,week_max))&(all_data.SV==sv)&(all_data.Comps==comp)].pivot_table(
                index='ProductTypeName',columns='Year',
                values=['SalesMny','SalesQty'],aggfunc='sum',fill_value=0).reset_index()
            item23.columns=column_names
            item23=item23.iloc[:,[0,2,1,4,3]]
            item23['Sales_Growth%']=item23['Sales_24']/item23['Sales_23']-1
            item23['Unit_Growth%']=item23['Units_Sold_24']/item23['Units_Sold_23']-1
            item23=item23.iloc[:,[0,1,2,5,3,4,6]]
            item23.replace([np.inf,-np.inf],inplace=True)
            item23.fillna(0,inplace=True)
        if state:
            item23=all_data[(all_data.State.isin(state))&(all_data.Week.between(week_min,week_max))&(all_data.SV==sv)&(all_data.Comps==comp)].pivot_table(
                index='ProductTypeName',columns='Year',
                values=['SalesMny','SalesQty'],aggfunc='sum',fill_value=0).reset_index()
            item23.columns=column_names
            item23=item23.iloc[:,[0,2,1,4,3]]
            item23['Sales_Growth%']=item23['Sales_24']/item23['Sales_23']-1
            item23['Unit_Growth%']=item23['Units_Sold_24']/item23['Units_Sold_23']-1
            item23=item23.iloc[:,[0,1,2,5,3,4,6]]
            item23.replace([np.inf,-np.inf],inplace=True)
            item23.fillna(0,inplace=True)
        if region:
            item23=all_data[(all_data.Region.isin(region))&(all_data.Week.between(week_min,week_max))&(all_data.SV==sv)&(all_data.Comps==comp)].pivot_table(
                index='ProductTypeName',columns='Year',
                values=['SalesMny','SalesQty'],aggfunc='sum',fill_value=0).reset_index()
            item23.columns=column_names
            item23=item23.iloc[:,[0,2,1,4,3]]
            item23['Sales_Growth%']=item23['Sales_24']/item23['Sales_23']-1
            item23['Unit_Growth%']=item23['Units_Sold_24']/item23['Units_Sold_23']-1
            item23=item23.iloc[:,[0,1,2,5,3,4,6]]
            item23.replace([np.inf,-np.inf],inplace=True)
            item23.fillna(0,inplace=True)
    return(item23)
```

```python
def sales_gc(coop=[],fc=[],dma=[],week_min=1,week_max=all_data.Week.max(),state=[],sv='',region=[],comp=''):
    test=all_data.fillna('Blank').groupby(['Year','Week','State','Region','CoOpName','SV','Comps','ShopID','FC','DMA'])['SalesMny'].sum().reset_index()\
    .merge(gc_total,on=['Year','Week','ShopID'],how='left').drop_duplicates(
    subset=['Year','Week','State','CoOpName','SV','Region','Comps','ShopID','FC','DMA'],keep='first')
    test.GC.fillna(0,inplace=True)
    if coop:
        test_piv=test[(test.CoOpName.isin(coop))&(test.SV==sv)&(test.Week.between(week_min,week_max))&(test.Comps==comp)]\
        .pivot_table(columns='Year',values=['SalesMny','GC'],aggfunc='sum')
        test_piv=test_piv[[2024,2023]]
        test_piv['Growth']=test_piv.iloc[:,0]/test_piv.iloc[:,1]-1
    if state:
        test_piv=test[(test.State.isin(state))&(test.SV==sv)&(test.Week.between(week_min,week_max))&(test.Comps==comp)]\
        .pivot_table(columns='Year',values=['SalesMny','GC'],aggfunc='sum')
        test_piv=test_piv[[2024,2023]]
        test_piv['Growth']=test_piv.iloc[:,0]/test_piv.iloc[:,1]-1
    if region:
        test_piv=test[(test.Region.isin(region))&(test.SV==sv)&(test.Week.between(week_min,week_max))&(test.Comps==comp)]\
        .pivot_table(columns='Year',values=['SalesMny','GC'],aggfunc='sum')
        test_piv=test_piv[[2024,2023]]
        test_piv['Growth']=test_piv.iloc[:,0]/test_piv.iloc[:,1]-1
    if fc:
        test_piv=test[(test.FC.isin(fc))&(test.SV==sv)&(test.Week.between(week_min,week_max))&(test.Comps==comp)]\
        .pivot_table(columns='Year',values=['SalesMny','GC'],aggfunc='sum')
        test_piv=test_piv[[2024,2023]]
        test_piv['Growth']=test_piv.iloc[:,0]/test_piv.iloc[:,1]-1
    if dma:
        test_piv=test[(test.DMA.isin(dma))&(test.SV==sv)&(test.Week.between(week_min,week_max))&(test.Comps==comp)]\
        .pivot_table(columns='Year',values=['SalesMny','GC'],aggfunc='sum')
        test_piv=test_piv[[2024,2023]]
        test_piv['Growth']=test_piv.iloc[:,0]/test_piv.iloc[:,1]-1
        
    return(test_piv)
        
```
