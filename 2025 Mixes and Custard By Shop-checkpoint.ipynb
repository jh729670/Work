```python
import xlsxwriter
with pd.ExcelWriter('2025 COGS thru ' + str(srs.EndDate.max().strftime('%m-%d-%y')) + '.xlsx', engine='xlsxwriter') as writer:
    sales.to_excel(writer, sheet_name='COGS', startrow=5, startcol=0, index=False)
    
    wb = writer.book
    ws = writer.sheets['COGS']
    
    # Define formats
    header_format = wb.add_format({'bold': True, 'italic': True})
    dollar_value = wb.add_format({'num_format': '$#,##0', 'align': 'center', 'valign': 'vcenter'})
    perc_value = wb.add_format({'num_format': '0.0%', 'align': 'center', 'valign': 'vcenter'})
    text_format = wb.add_format({'align': 'center', 'valign': 'vcenter'})
    border_format = wb.add_format({'right': 2, 'left': 2, 'bottom': 2, 'top': 2})
    
    # Row banding formats (for alternating colors)
    odd_row_format = wb.add_format({'bg_color': '#D3D3D3'})  # Light grey for odd rows
    even_row_format = wb.add_format({'bg_color': '#FFFFFF'})  # White for even rows
    
    # Apply border to all cells with data
    ws.conditional_format(xlsxwriter.utility.xl_range(5, 0, len(sales) + 5, len(sales.columns) - 1),
                          {'type': 'no_errors', 'format': border_format})
    
    # Apply alternating row colors only where there is data
    for row in range(5, len(sales) + 5):
        for col in range(0, len(sales.columns)):
            cell_value = sales.iloc[row - 5, col]  # Get the cell value
            if pd.notna(cell_value):  # Check if the cell has data
                if (row - 5) % 2 == 0:  # Even rows relative to data
                    ws.write(row, col, cell_value, even_row_format)
                else:
                    ws.write(row, col, cell_value, odd_row_format)
                    if (row - 5) % 2 == 0:  # Even rows
                        ws.conditional_format(row, col, row, col, {'type': 'no_errors', 'format': even_row_format})
                    else:  # Odd rows
                        ws.conditional_format(row, col, row, col, {'type': 'no_errors', 'format': odd_row_format})
                    
    # Write header text
    text1 = 'Ritas Franchise Operations,LLC'
    text2 = 'Cost of Goods Sold By Category'
    text3 = 'Orders thru ' + str(invoice.InvoiceDate.max().strftime('%m/%d/%y'))
    text4 = 'Sales thru ' + str(srs.EndDate.max())
    
    ws.write(0, 1, text1, header_format)
    ws.write(1, 1, text2, header_format)
    ws.write(2, 1, text3, header_format)
    ws.write(3, 1, text4, header_format)
    
    # Adjust column widths and apply formats
    ws.set_column('A:D', 20, text_format)
    ws.set_column('B:B', 45, text_format)
    ws.set_column('E:F', 17, dollar_value)
    ws.set_column('H:H', 16, dollar_value)
    ws.set_column('J:J', 24, dollar_value)
    ws.set_column('G:G', 15, perc_value)
    ws.set_column('I:I', 15, perc_value)
    ws.set_column('K:K', 15, perc_value)
    
    # Insert image and hide gridlines
    ws.insert_image('A1', 'Logo.png')
    ws.hide_gridlines(2)
    
    # Enable autofilter for the data range
    ws.autofilter(5, 0, sales.shape[0], sales.shape[1] - 1)
```
