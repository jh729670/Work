```python
import pandas as pd
import pandas_datareader as pdr
import datetime as dt
from datetime import date
import numpy as np
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
ticks=['SOL-USD','BTC-USD','NOS-USD','RENDER-USD','JUP-USD','RAY-USD','PYTH-USD','HNT-USD','JTO-USD','ORCA-USD']
df=yf.Tickers(ticks).history(period='6mo').Close
```

```python
df=df.drop('RENDER-USD',axis=1)
```

```python
df
```

```python
#Creates corrleation dataframe 
df.corr()
```

```python
#Plots the correlation dateframe as a heatmap to easily view highly correlated assets

figure(figsize=(12,10),dpi=80)
sns.heatmap(df.corr(),annot=True).set(title='Heat Map of Crypto Correlation to Solana');
```

```python
sol_pairs=df.div(df['SOL-USD'],axis=0)
```

```python
# Assuming rolling_5 and rolling_20 are already computed
rolling_5 = df.rolling(5).mean()
rolling_20 = df.rolling(20).mean()

# Loop through columns and plot the rolling averages
for index, column in list(enumerate(df.columns)):
    fig,ax=plt.subplots(figsize=(15,10))
    #plt.figure(figsize=(10, 6))  # Create a new figure for each column
    sns.lineplot(x=df.index,y=df[column],label=f'{column} Price',color='black')
    sns.lineplot(x=df.index, y=rolling_5[column], label=f'Rolling 5 - {column}')
    sns.lineplot(x=df.index, y=rolling_20[column], label=f'Rolling 20 - {column}')
    ax.set_title(str(column)+' 5 and 20 Day Rolling Averages',fontsize=20)
    plt.xlabel('Dates')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

```

```python
for index, column in list(enumerate(sol_pairs.columns)):
    fig,ax=plt.subplots(figsize=(15,10))
    #plt.figure(figsize=(10, 6))  # Create a new figure for each column
    sns.lineplot(x=sol_pairs.index, y=sol_pairs[column], label=f'SOL-USD - {column} Pair')
    ax.set_title(str(column)+'/SOL Pair',fontsize=20)
    upper_limit=sol_pairs.mean()+2*sol_pairs.std()
    up_limit=sol_pairs.mean()+sol_pairs.std()
    mean=sol_pairs.mean()
    low_limit=sol_pairs.mean()-sol_pairs.std()
    lower_limit=sol_pairs.mean()-2*sol_pairs.std()
    plt.axhline(upper_limit[column],color='green')
    plt.axhline(up_limit[column],color='green')
    plt.axhline(mean[column],color='black')
    plt.axhline(low_limit[column],color='red')
    plt.axhline(lower_limit[column],color='red')
    plt.xlabel('Dates')
    plt.ylabel('Price')
    plt.legend()
    plt.show()
```
