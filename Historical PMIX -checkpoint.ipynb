```python
def natalie_sales(dma=[],coop=[],msa='',shopid=0,week_min=1,week_max=data.Week.max(),state=''):
    if shopid >0:
        sales=data[(data.ShopID==shopid)&(data.Week.between(week_min,week_max))].SalesMny.sum()
    elif shopid==0:
        if dma:
            sales=data[(data.DMA.isin(dma))&(data.Week.between(week_min,week_max))].SalesMny.sum()
        if coop:
            sales=data[(data.CoOpName.isin(coop))&(data.Week.between(week_min,week_max))].SalesMny.sum()
        if msa:
            sales=data[(data.MSAName.str.contains(msa))&(data.Week.between(week_min,week_max))].SalesMny.sum()
        if state:
            sales=data[(data.State==state)&(data.Week.between(week_min,week_max))].SalesMny.sum()
    return(sales)
```

```python
def growth(week_min=1,week_max=all_data.Week.max(),sv='',comp=''):
        column_names=['ProductTypeName','Sales_23','Sales_24','Units_Sold_23','Units_Sold_24']
        item23=all_data[(all_data.Week.between(week_min,week_max))&(all_data.SV==sv)&(all_data.Comps==comp)].pivot_table(index='ProductTypeName',columns='Year',
                                               values=['SalesMny','SalesQty'],aggfunc='sum',fill_value=0).reset_index()
        item23.columns=column_names
        item23=item23.iloc[:,[0,2,1,4,3]]
        item23['Sales_Growth%']=item23['Sales_24']/item23['Sales_23']-1
        item23['Unit_Growth%']=item23['Units_Sold_24']/item23['Units_Sold_23']-1
        item23=item23.iloc[:,[0,1,2,5,3,4,6]]
        item23.replace([np.inf,-np.inf],inplace=True)
        item23.fillna(0,inplace=True)
        return(item23)
```

```python
def custard_pmix(state='',week_min=0):
    if state:
        sales=year_round[(year_round.State==state)&(year_round.Week.isin(week_min))].SalesMny.sum()
        qty=year_round[(year_round.State==state)&(year_round.Week.isin(week_min))].SalesQty.sum()
        item23=year_round[(year_round.State==state)&(year_round.Week.isin(week_min))].pivot_table(
                index='ProductTypeName',
                values=['SalesMny','SalesQty'],aggfunc='sum').reset_index()
        item23['Sales%']=item23.SalesMny/sales
        item23['Qty%']=item23.SalesQty/qty
    else:
        sales=year_round[year_round.Week.isin(week_min)].SalesMny.sum()
        qty=year_round[year_round.Week.isin(week_min)].SalesQty.sum()
        item23=year_round[year_round.Week.isin(week_min)].pivot_table(
                index='ProductTypeName',
                values=['SalesMny','SalesQty'],aggfunc='sum').reset_index()
        item23['Sales%']=item23.SalesMny/sales
        item23['Qty%']=item23.SalesQty/qty
    return(item23)
```
