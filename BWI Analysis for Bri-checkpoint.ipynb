# For Natatlie
## Product Mix for Florida for February and March

```python
florida=data[data.State=='FL']
florida['Month']=florida.StartDate.dt.month
florida['Year']=florida.StartDate.dt.year
```

```python
florida
```

```python
florida2=data2[data2.State=="FL"]
florida2['Month']=florida2.StartDate.dt.month
florida2['Year']=florida2.StartDate.dt.year
```

```python
florida2
```

```python
df=pd.concat([florida, florida2])
```

```python
florida_total=round(df.groupby(['Year','Month','ProductTypeName'])[['SalesQty','SalesMny']].sum(),2).reset_index()
```

```python
florida_total
```

```python
round(df.groupby(['Year','Month','ShopID','ShopName','ProductTypeName'])[['SalesQty','SalesMny']].sum(),2).reset_index()
```

```python
writer2=pd.ExcelWriter('Florida Product Mix 12.12.22.xlsx')
florida_total.to_excel(writer2,sheet_name='Raw Total',index=False)
round(df.groupby(['Year','Month','ShopID','ShopName','ProductTypeName'])[['SalesQty','SalesMny']].sum(),2).reset_index().to_excel(writer2,sheet_name='Raw Shop Total',index=False)
writer2.save()
writer2.close()
```
