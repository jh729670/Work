```python
may_22=df_merge[df_merge.dteDate=='2022-05-07'].groupby(['intStoreID','intRitasStoreID'])[['intQuantity','mnyPosSales']].sum().reset_index()
may_22=may_22.merge(data3[(data3.dteDate=='2022-05-07')&(data3.intZTapeID==6)][['intStoreID','vchValue']],on='intStoreID',how='left')
qty=df_merge[(df_merge.dteDate=='2022-05-07')&(df_merge.vchProductTypeName=='Gelati')].pivot_table(index='intRitasStoreID',
                                                                                               columns='vchProductSizeName',
                                                                                               values='intQuantity',aggfunc='sum',fill_value=0).reset_index().drop(['12 oz Rainbow Gelati',
                                                                                               '5-Layer Gelati'],axis=1)
sales=df_merge[(df_merge.dteDate=='2022-05-07')&(df_merge.vchProductTypeName=='Gelati')].pivot_table(index='intRitasStoreID',
                                                                                               columns='vchProductSizeName',
                                                                                               values='mnyPosSales',aggfunc='sum',fill_value=0).reset_index().drop(['12 oz Rainbow Gelati',
                                                                                               '5-Layer Gelati'],axis=1)
sales_m=df_merge[(df_merge.dteDate=='2022-05-07')&(df_merge.vchProductTypeName=='Misto')].pivot_table(index='intRitasStoreID',
                                                                                               columns='vchProductSizeName',
                                                                                               values='mnyPosSales',aggfunc='sum',fill_value=0).reset_index()
qty_m=df_merge[(df_merge.dteDate=='2022-05-07')&(df_merge.vchProductTypeName=='Misto')].pivot_table(index='intRitasStoreID',
                                                                                               columns='vchProductSizeName',
                                                                                               values='intQuantity',aggfunc='sum',fill_value=0).reset_index()
may_22=may_22.merge(qty,on='intRitasStoreID',how='left')
may_22=may_22.merge(sales,on='intRitasStoreID',how='left')
may_22=may_22.merge(qty_m,on='intRitasStoreID',how='left')
may_22=may_22.merge(sales_m,on='intRitasStoreID',how='left')

may_22.rename(columns={'intRitasStoreID':'ShopID','intQuantity':'Quantity','mnyPosSales':'Sales',
                      'vchValue':'Guest_Count','Large Gelati_x':'Lg_Qty_Gelati','Small Gelati_x':'Sm_Qty_Gelati',
                      'Large Gelati_y':'Lg_Sales_Gelati','Small Gelati_y':'Sm_Sales_Gelati','Small Misto_x':'Sm_Qty_Misto',
                      'Large Misto_x':'Lg_Qty_Misto','Large Misto_y':'Lg_Sales_Misto','Small Misto_y':'Sm_Sales_Misto'},inplace=True)
may_22
```
